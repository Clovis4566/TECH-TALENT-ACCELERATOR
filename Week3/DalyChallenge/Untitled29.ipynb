```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuration pour la reproductibilité
np.random.seed(42)

# Liste des villes et mois
villes = [f"Ville_{i+1}" for i in range(10)]
mois = ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Juin', 'Juil', 'Aoû', 'Sep', 'Oct', 'Nov', 'Déc']

# Génération des données : 10 villes x 12 mois, températures entre -5 et 35
data = np.random.uniform(-5, 35, size=(10, 12))

# Création du DataFrame
df_meteo = pd.DataFrame(data, index=villes, columns=mois)

print("--- Aperçu des données ---")
print(df_meteo.head())
# Calcul de la moyenne annuelle par ville (moyenne sur les colonnes, donc axis=1)
moyennes_annuelles = df_meteo.mean(axis=1)

# Identification des villes extrêmes
ville_max = moyennes_annuelles.idxmax()
ville_min = moyennes_annuelles.idxmin()

print("\n--- Analyse des températures ---")
print(f"La ville la plus chaude en moyenne est : {ville_max} ({moyennes_annuelles[ville_max]:.2f}°C)")
print(f"La ville la plus froide en moyenne est : {ville_min} ({moyennes_annuelles[ville_min]:.2f}°C)")
# Configuration du graphique
plt.figure(figsize=(12, 6))

# Visualisation des tendances mensuelles
for ville in df_meteo.index:
    plt.plot(df_meteo.columns, df_meteo.loc[ville], marker='o', label=ville, alpha=0.6)

plt.title("Évolution des températures mensuelles par ville")
plt.xlabel("Mois")
plt.ylabel("Température (°C)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualisation des moyennes annuelles
plt.figure(figsize=(10, 5))
moyennes_annuelles.sort_values().plot(kind='bar', color='skyblue')
plt.title("Température moyenne annuelle par ville")
plt.ylabel("Température moyenne (°C)")
plt.xticks(rotation=45)
plt.show()
```
