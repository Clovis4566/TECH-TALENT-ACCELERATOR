```python
#exercice 2
import numpy as np

# 1. Génération d'un tableau de 50 nombres aléatoires (ici entre 1 et 100)
# npy.random.randint(min, max, quantité)
donnees = np.random.randint(1, 101, 50)

print("--- Les 50 nombres générés ---")
print(donnees)
print("-" * 40)

# 2. Calcul de la moyenne
moyenne = np.mean(donnees)

# 3. Calcul de la médiane
mediane = np.median(donnees)

# 4. Calcul de l'écart type
ecart_type = np.std(donnees)

# Affichage des résultats
print(f"Moyenne    : {moyenne:.2f}")
print(f"Médiane    : {mediane:.2f}")
print(f"Écart type : {ecart_type:.2f}")
```

```python
import numpy as np
from datetime import datetime

# 1. Création du tableau de dates pour le mois de janvier 2023
# On définit le début (inclus) et la fin (exclue, donc on met le 1er février) avec un pas de 1 jour ('D')
dates_janvier = np.arange('2023-01-01', '2023-02-01', dtype='datetime64[D]')

print("--- Dates d'origine (Format par défaut ISO YYYY-MM-DD) ---")
print(dates_janvier[:5]) # On affiche les 5 premières pour l'exemple
print(f"Total de jours générés : {len(dates_janvier)}\n")
print("-" * 50)

# 2. Conversion dans un autre format (ex: AAAA/MM/JJ)
# NumPy stocke les dates de façon brute. Pour changer le format d'affichage (comme remplacer les tirets par des slashs),
# on utilise le module 'datetime' intégré à Python ou l'outil de gestion des chaînes de NumPy.
# The `datetime_as_string` function does not accept a 'format' argument for custom formatting.
# We need to convert to Python datetime objects and then use strftime.
dates_formatees = [np.datetime64(d).astype(datetime).strftime('%Y/%m/%d') for d in dates_janvier]
dates_formatees = np.array(dates_formatees) # Convert back to numpy array for consistency if desired

print("--- Dates converties (Format AAAA/MM/JJ) ---")
print(dates_formatees[:5]) # Affichage des 5 premières dates converties
```

```python
import numpy as np
import pandas as pd

# 1. Création du DataFrame avec des nombres aléatoires
# On génère une matrice NumPy de 10 lignes et 3 colonnes avec des notes/valeurs entre 0 et 100
donnees_aleatoires = np.random.randint(0, 101, size=(10, 3))

# On transforme cette matrice en DataFrame Pandas avec des noms de colonnes
df = pd.DataFrame(donnees_aleatoires, columns=['Maths', 'Physique', 'Code'])

print("--- DataFrame d'origine (Les 10 élèves) ---")
print(df)
print("\n" + "="*40 + "\n")

# 2. Sélection conditionnelle des données
# Exemple : On veut filtrer et afficher uniquement les lignes où la note de 'Maths' est supérieure ou égale à 50
print("--- Sélection conditionnelle (Maths >= 50) ---")
filtre_maths = df[df['Maths'] >= 50]
print(filtre_maths)
print("\n" + "="*40 + "\n")

# 3. Fonctions d'agrégation (Somme et Moyenne)
print("--- Fonctions d'agrégation ---")

# Somme totale des points par matière (colonne par colonne)
somme_colonnes = df.sum()
print("Somme totale par matière :")
print(somme_colonnes)
print()

# Moyenne de chaque matière
moyenne_colonnes = df.mean()
print("Moyenne générale par matière :")
print(moyenne_colonnes)
```

```python
import numpy as npy

# Fixer la graine aléatoire pour avoir les mêmes résultats à chaque exécution
np.random.seed(42)

# Génération des données de l'énoncé
productivity_before = npy.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# 1. Calcul des moyennes avant et après
moyenne_avant = np.mean(productivity_before)
moyenne_apres = np.mean(productivity_after)

print(f"Moyenne de productivité AVANT : {moyenne_avant:.2f}")
print(f"Moyenne de productivité APRÈS : {moyenne_apres:.2f}")
print("-" * 50)

# 2. Calcul de la différence individuelle pour chaque employé
differences = productivity_after - productivity_before
gain_moyen = np.mean(differences)

print(f"Gain moyen de productivité par employé : +{gain_moyen:.2f}")

# 3. Test de l'hypothèse : Quel pourcentage d'employés a réellement progressé ?
progression_positive = differences > 0
pourcentage_succes = npy.sum(progression_positive) / len(differences) * 100

print(f"Pourcentage d'employés ayant amélioré leur productivité : {pourcentage_succes:.1f}%")

# Conclusion basée sur les données NumPy
if gain_moyen > 0 and pourcentage_succes > 50:
    print("\n👉 Conclusion : L'hypothèse nulle (H0) est rejetée empiriquement. Le programme de formation semble efficace ! (Gain moyen positif et majorité d'employés en progression).")
else:
    print("\n👉 Conclusion : Impossible de rejeter l'hypothèse nulle. La formation n'a pas prouvé son efficacité.")
```

```python
import numpy as npy
import pandas as pd

# Fixer la graine aléatoire pour la reproductibilité
npy.random.seed(2023)

# 1. Génération d'une séquence de dates pour toute l'année 2023 (fréquence journalière 'D')
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# 2. Génération de données aléatoires pour simuler une métrique (ex: ventes quotidiennes)
valeurs_ventes = npy.random.randint(100, 500, size=len(dates))

# 3. Création du DataFrame avec les dates en guise d'INDEX (Crucial pour le découpage temporel)
ts_data = pd.DataFrame(data={'Ventes': valeurs_ventes}, index=dates)

print(f"--- Série temporelle créée ---")
print(f"Nombre total de jours : {len(ts_data)}")
print("Aperçu des premières lignes :")
print(ts_data.head(3))
print("\n" + "="*50 + "\n")


# ==========================================
# DÉMONSTRATION DU DÉCOUPAGE (SLICING)
# ==========================================

# Intervalle 1 : De janvier à mars (Trimestre 1)
t1 = ts_data['2023-01':'2023-03']
print(f"🛒 Janvier à Mars : {len(t1)} jours | Somme des ventes = {t1['Ventes'].sum()}")

# Intervalle 2 : Avril à juin (Trimestre 2)
t2 = ts_data['2023-04':'2023-06']
print(f"🌱 Avril à Juin   : {len(t2)} jours | Somme des ventes = {t2['Ventes'].sum()}")

# Intervalle 3 : De juillet à septembre (Trimestre 3)
t3 = ts_data['2023-07':'2023-09']
print(f"☀️ Juillet à Sept. : {len(t3)} jours | Somme des ventes = {t3['Ventes'].sum()}")

# Intervalle 4 : D'octobre à décembre (Trimestre 4)
t4 = ts_data['2023-10':'2023-12']
print(f"🍂 Octobre à Déc.  : {len(t4)} jours | Somme des ventes = {t4['Ventes'].sum()}")


print("\n" + "="*50 + "\n")
# Exemple bonus : Affichage des détails du premier trimestre pour illustrer le découpage
print("Détail des bornes du premier découpage (Janvier à Mars) :")
print(f"Première date : {t1.index.min().strftime('%Y-%m-%d')}")
print(f"Dernière date : {t1.index.max().strftime('%Y-%m-%d')}")
```

```python
import numpy as npy
import pandas as pd

print("=== 1. CONVERSION : NUMPY ➡️ PANDAS ===")

# Création d'un tableau NumPy 2D (3 lignes, 2 colonnes)
tableau_numpy = npy.array([
    [10, 20],
    [30, 40],
    [50, 60]
])

print("Type d'origine :", type(tableau_numpy))
print(tableau_numpy)
print()

# Conversion en DataFrame Pandas
# On passe le tableau en argument et on peut optionnellement nommer les colonnes
df_pandas = pd.DataFrame(tableau_numpy, columns=['Colonne_A', 'Colonne_B'])

print("Après conversion en Pandas :")
print("Nouveau type :", type(df_pandas))
print(df_pandas)


print("\n" + "="*50 + "\n")


print("=== 2. CONVERSION : PANDAS ➡️ NUMPY ===")

# Nous allons reprendre le DataFrame 'df_pandas' créé ci-dessus et le reconvertir
# La méthode moderne et recommandée est '.to_numpy()'
tableau_retour = df_pandas.to_numpy()

print("Après reconversion en NumPy :")
print("Nouveau type :", type(tableau_retour))
print(tableau_retour)
```

```python
import numpy as npy
import matplotlib.pyplot as plt

# 1. Génération des données avec NumPy
# On crée 50 points sur l'axe X (de 1 à 50)
x = npy.arange(1, 51)

# On génère 50 nombres aléatoires, et on utilise .cumsum() pour les additionner
# au fur et à mesure (cela crée une ligne continue qui monte et descend de façon réaliste)
npy.random.seed(42)  # Pour avoir la même courbe à chaque lancer
y = npy.random.normal(loc=2, scale=5, size=50).cumsum()

# 2. Création et personnalisation du graphique avec Matplotlib
plt.figure(figsize=(10, 5)) # Définit la taille de la fenêtre du graphique

# On trace la ligne : x en abscisse, y en ordonnée
# On ajoute des options de style : couleur bleue, épaisseur de ligne (lw), et style de marqueurs
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o', label='Évolution de la valeur')

# 3. Habillage du graphique (Titres et légendes)
plt.title("Visualisation de base : Graphique linéaire avec NumPy & Matplotlib", fontsize=14, fontweight='bold')
plt.xlabel("Temps / Étapes (Axe X)", fontsize=12)
plt.ylabel("Valeurs Aléatoires Cumulées (Axe Y)", fontsize=12)

# Ajout d'une grille en arrière-plan pour mieux lire les valeurs
plt.grid(True, linestyle='--', alpha=0.6)

# Affichage de la légende (basée sur le 'label' défini dans plt.plot)
plt.legend()

# 4. Commande finale pour afficher le graphique à l'écran
plt.show()
```
