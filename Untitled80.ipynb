## Daily Challenge: Fine-tuning LLMs with LoRA

First, let's install the necessary libraries: `peft` (version 0.4.0) and `datasets`.

```python
# Install necessary libraries
%pip install peft==0.4.0 datasets transformers accelerate bitsandbytes
```

Next, we'll create a cache directory to store model and dataset downloads, and then import all the required modules.

```python
# Create a cache directory
import os
if not os.path.exists('cache'):
    os.makedirs('cache')

# Import necessary libraries
from datasets import load_dataset
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from peft import LoraConfig, get_peft_model, PeftModel
import torch
import time
```

Now, we'll load the pre-trained model `bigscience/bloomz-560m` and its tokenizer. We'll also ensure the tokenizer has a `pad_token` set, as it's crucial for batch processing during training.

```python
# Model and Tokenizer
model_name = "bigscience/bloomz-560m"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Set pad_token_id if not already set
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    # Also update the model's config for padding token id
    # foundation_model.config.pad_token_id = tokenizer.eos_token_id

foundation_model = AutoModelForCausalLM.from_pretrained(model_name)
```

We will now load the `Abirate/english_quotes` dataset, take a 10% random sample of the training split, and tokenize the `quote` column. This prepares the data for training the language model.

```python
# Load the dataset
data = load_dataset("Abirate/english_quotes")

# Sample 10% of the training split
train_data = data["train"].shuffle(seed=42).select(range(int(len(data["train"]) * 0.1)))

# Preprocess the data by tokenizing the 'quote' column
def tokenize_function(examples):
    return tokenizer(examples["quote"], truncation=True, max_length=128)

tokenized_data = train_data.map(tokenize_function, batched=True, remove_columns=["quote", "author"]) # Remove original columns

# The Trainer expects a 'labels' column, which can be the same as 'input_ids' for causal language modeling
tokenized_data = tokenized_data.rename_column("input_ids", "labels")

display(tokenized_data)
```

Now, we'll configure LoRA using `LoraConfig` with the specified parameters. `target_modules` is set to `query_key_value`, which is appropriate for BLOOM models. We will then apply these adapter layers to our foundation model using `get_peft_model`.

```python
# Configure LoRA
lora_config = LoraConfig(
    r=1,
    lora_alpha=1,
    target_modules=["query_key_value"], # adapted for BLOOM
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

# Add the adapter layers to the foundation model
peft_model = get_peft_model(foundation_model, lora_config)
print(peft_model.print_trainable_parameters())
```

We will now set up the `TrainingArguments` and initialize the `Trainer` for fine-tuning. We'll use a data collator specifically for language modeling, which will dynamically pad sequences to the longest length in a batch.

```python
# Set up Training Arguments
output_directory = os.path.join("../cache/working", "peft_lab_outputs")

training_args = TrainingArguments(
    report_to="none",
    output_dir=output_directory,
    auto_find_batch_size=True, # Automatically finds a suitable batch size
    learning_rate=3e-2,
    num_train_epochs=1, # Set to 1 epoch for quick demonstration on CPU
    use_cpu=True, # Explicitly use CPU
    save_strategy="epoch", # Save checkpoints at the end of each epoch
    logging_dir=f"{output_directory}/logs", # Directory for storing logs
    logging_strategy="steps", # Log every N steps
    logging_steps=10,
)

# Initialize and train the model
trainer = Trainer(
    model=peft_model,
    args=training_args,
    train_dataset=tokenized_data,
    data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False)
)

trainer.train()
```

After training, we will save the fine-tuned LoRA model to a unique directory named with a timestamp, ensuring we can easily retrieve this specific version later.

```python
# Save the fine-tuned LoRA model
time_now = int(time.time())
peft_model_path = os.path.join(output_directory, f"peft_model_{time_now}")
trainer.model.save_pretrained(peft_model_path)

print(f"Fine-tuned PEFT model saved to: {peft_model_path}")
```

Finally, we will load the base model, and then apply the saved LoRA adapter using `PeftModel.from_pretrained`. We set `is_trainable=False` as we only intend to use the model for inference. Then, we will generate text based on the prompt: "Two things are infinite: ".

```python
# Load the base model again for inference
inference_model = AutoModelForCausalLM.from_pretrained(model_name)

# Load the PEFT model using pre-defined LoRA configs and foundation model
# We set `is_trainable=False` to avoid further training.
peft_inference_model = PeftModel.from_pretrained(inference_model, peft_model_path, is_trainable=False)

# Generate output tokens
inputs = tokenizer("Two things are infinite: ", return_tensors="pt")

# Move inputs to device if using GPU, otherwise it stays on CPU.
# For CPU, it's already on the correct device.

# Generate text
outputs = peft_inference_model.generate(
    input_ids=inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_new_tokens=50, # Generate up to 50 new tokens
    num_beams=5, # Use beam search for better quality
    early_stopping=True,
    pad_token_id=tokenizer.pad_token_id # Ensure pad_token_id is set for generation
)

print("Generated text:")
print(tokenizer.batch_decode(outputs, skip_special_tokens=True)[0])
```
