## Reconnaissance de chiffres manuscrits avec un réseau neuronal entièrement connecté

### 1. Charger et prétraiter l'ensemble de données MNIST

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

# Charger le jeu de données MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normaliser les valeurs des pixels de l'image pour qu'elles soient comprises entre 0 et 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convertir les étiquettes au format encodé one-hot
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

print(f"Forme de l'ensemble d'entraînement: {x_train.shape}, {y_train.shape}")
print(f"Forme de l'ensemble de test: {x_test.shape}, {y_test.shape}")
```

```python
# Afficher des exemples d'images avec leurs étiquettes correspondantes
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"Label: {np.argmax(y_train[i])}")
    plt.axis('off')
plt.suptitle("Exemples d'images MNIST et leurs étiquettes")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### 2. Construire un réseau neuronal entièrement connecté

```python
import tensorflow as tf
from tensorflow import keras

# Définir un modèle séquentiel à l'aide de Keras
model = keras.Sequential([
    # Aplatir les images d'entrée 28x28 en un seul vecteur (28*28 = 784)
    keras.layers.Flatten(input_shape=(28, 28)),
    # Ajoutez deux couches cachées avec activation ReLU
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    # Ajoutez une couche de sortie avec activation Softmax pour la classification multiclasse
    keras.layers.Dense(10, activation='softmax') # 10 classes pour les chiffres 0-9
])

# Compilez le modèle en utilisant l'entropie croisée catégorielle comme fonction de perte et la précision comme métrique d'évaluation.
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

### 3. Entraîner le réseau neuronal

```python
# Ajuster le modèle aux données d'entraînement pendant 10 époques.
# Utilisez un ensemble de validation pour suivre les performances pendant l'entraînement.
history = model.fit(x_train, y_train, epochs=10, validation_split=0.2)

# Observez les tendances en matière de perte et de précision au fil des époques.
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Précision d\'entraînement')
plt.plot(history.history['val_accuracy'], label='Précision de validation')
plt.xlabel('Époque')
plt.ylabel('Précision')
plt.legend()
plt.title('Précision du modèle au fil des époques')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Perte d\'entraînement')
plt.plot(history.history['val_loss'], label='Perte de validation')
plt.xlabel('Époque')
plt.ylabel('Perte')
plt.legend()
plt.title('Perte du modèle au fil des époques')

plt.tight_layout()
plt.show()
```

### 4. Évaluer les performances du modèle

```python
# Calculer la précision sur l'ensemble de données de test
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Précision sur l'ensemble de test: {accuracy:.4f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Obtenir les prédictions du modèle sur l'ensemble de test
y_pred_probs = model.predict(x_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

# Afficher une matrice de confusion pour les chiffres mal classés
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Prédiction')
plt.ylabel('Vraie étiquette')
plt.title('Matrice de Confusion')
plt.show()
```

```python
# Identifier les chiffres avec lesquels le modèle a le plus de difficultés.
# Calculer le nombre de classifications erronées pour chaque classe
incorrect_predictions = y_true[y_true != y_pred]
incorrect_predicted = y_pred[y_true != y_pred]

# Compter les erreurs par classe vraie
error_counts = np.bincount(incorrect_predictions, minlength=10)

print("Nombre de classifications erronées par chiffre (vraie étiquette) :")
for i, count in enumerate(error_counts):
    print(f"Chiffre {i}: {count} erreurs")

# Chiffres avec lesquels le modèle a le plus de difficultés (ceux avec le plus d'erreurs)
most_difficult_digits = np.argsort(error_counts)[::-1]

print(f"\nLes chiffres avec lesquels le modèle a le plus de difficultés sont : {most_difficult_digits[0]}, {most_difficult_digits[1]}, etc.")

# Visualisation de quelques erreurs
print("\nExemples de classifications erronées :")
plt.figure(figsize=(10, 5))
errors_found = 0
for i in range(len(y_true)):
    if y_true[i] != y_pred[i]:
        plt.subplot(2, 5, errors_found + 1)
        plt.imshow(x_test[i], cmap='gray')
        plt.title(f"Vrai: {y_true[i]}, Prédit: {y_pred[i]}")
        plt.axis('off')
        errors_found += 1
        if errors_found >= 10: # Afficher jusqu'à 10 erreurs
            break
plt.suptitle('Exemples de classifications erronées')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
