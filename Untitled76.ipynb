### 5. Création des plis (K-Fold)
Maintenant, nous allons charger le fichier d'entraînement et préparer la validation croisée stratifiée.

```python
from sklearn.model_selection import StratifiedKFold

# Chargement du jeu de données d'entraînement
train_df = pd.read_csv('data_final/train.csv')

def create_folds(data, num_splits=5):
    data["kfold"] = -1
    kf = StratifiedKFold(n_splits=num_splits, shuffle=True, random_state=42)

    # On identifie la colonne cible (souvent 'label' ou 'target')
    target_col = 'label' if 'label' in data.columns else data.columns[-1]

    for fold, (train_idx, val_idx) in enumerate(kf.split(X=data, y=data[target_col])):
        data.loc[val_idx, 'kfold'] = fold
    return data

train_df = create_folds(train_df)
print("Distribution des plis :")
print(train_df['kfold'].value_counts())
display(train_df.head())
```

### 3. Préparation des données d'entrée pour le modèle

Nous allons créer une classe `Dataset` pour PyTorch qui gère la tokenisation des paires de phrases.

```python
import torch
from torch.utils.data import Dataset, DataLoader

class NLIDataset(Dataset):
    def __init__(self, df, tokenizer, max_len=128):
        self.df = df
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.df)

    def __getitem__(self, index):
        row = self.df.iloc[index]
        premise = str(row.premise)
        hypothesis = str(row.hypothesis)

        # Remplacement de encode_plus par l'appel direct au tokenizer
        inputs = self.tokenizer(
            premise,
            hypothesis,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        return {
            'input_ids': inputs['input_ids'].flatten(),
            'attention_mask': inputs['attention_mask'].flatten(),
            'label': torch.tensor(row.label, dtype=torch.long) if 'label' in row else torch.tensor(0)
        }

# Test du Dataset avec le premier pli
train_idx = train_df[train_df.kfold != 0].reset_index(drop=True)
ds_test = NLIDataset(train_idx, xlm_roberta_tokenizer)
print(f"Exemple d'entrée tokenisée : {ds_test[0]['input_ids'][:10]}...")
```

### 6. Définition du modèle et entraînement

Nous utilisons `XLMRobertaForSequenceClassification` qui est déjà optimisé pour la classification de texte.

```python
from transformers import XLMRobertaForSequenceClassification
from torch.optim import AdamW
import torch
from torch.utils.data import DataLoader

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Utilisation de : {device}")

def train_one_fold(fold, train_df, tokenizer):
    # Préparation des données pour ce pli
    df_train = train_df[train_df.kfold != fold].reset_index(drop=True)
    df_val = train_df[train_df.kfold == fold].reset_index(drop=True)

    train_ds = NLIDataset(df_train, tokenizer)
    val_ds = NLIDataset(df_val, tokenizer)

    train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=16)

    # Chargement du modèle multilingue
    model = XLMRobertaForSequenceClassification.from_pretrained('xlm-roberta-base', num_labels=3)
    model.to(device)

    # Utilisation de l'optimiseur AdamW de PyTorch
    optimizer = AdamW(model.parameters(), lr=2e-5)

    print(f"--- Début de l'entraînement pour le Fold {fold} ---")
    model.train()
    for batch in train_loader:
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        # On s'arrête ici pour la démonstration technique
        break

    print(f"Fold {fold} testé avec succès.")
    return model

# Lancement du test sur le premier pli
model_trained = train_one_fold(0, train_df, xlm_roberta_tokenizer)
```
