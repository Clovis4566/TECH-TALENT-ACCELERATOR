## Exercise 1: Rewrite and Optimize a Vague Prompt

### Tasks
- Identify 3 critical issues with the prompt.
- Rewrite it using: role prompting, clear instructions, and format/tone/length constraints.
- Add a final line to make it reusable for future social posts.

```python
critical_issues = [
    'The prompt is too vague; it doesn\'t specify the angle or type of productivity tips desired.',
    'No target audience is defined, so the tone and content cannot be tailored effectively.',
    'The prompt lacks any mention of the product (FlowNest) or a call to action for the post.',
]
critical_issues
```

## Exercise 2: Multi-Part Prompt for Quiz Generation
**Scenario**: Create an educational prompt for 7th graders about volcanic eruptions.

### Tasks
- Ask for: 2-bullet summary + 3 MCQs (1 correct, 2 distractors).
- Require friendly tone, simple vocab (11?13 y/o), and Google Slides-ready output.
- Explain 3 ways your prompt improves on: `Make a quiz for kids about this article.`

```python
quiz_prompt = """As an educational content creator for 7th graders (ages 11-13), your task is to create a learning resource about volcanic eruptions. You will be provided with an article on volcanic eruptions. Your output should be structured as follows and delivered in a friendly tone with simple vocabulary:

1.  **2-Bullet Summary:** Provide a concise, easy-to-understand summary of the article in two bullet points.
2.  **3 Multiple-Choice Questions (MCQs):** Generate three multiple-choice questions based on the article. Each question must have one correct answer and two plausible, but incorrect, distractors. Format these questions for easy copy-pasting into Google Slides, clearly indicating the correct answer.

Ensure the entire output is engaging for a 7th-grade audience and directly uses information from the provided article without introducing new concepts. This prompt is designed to be reusable for creating similar educational content for other topics.

**Article on Volcanic Eruptions:** [INSERT ARTICLE TEXT HERE]"""
quiz_prompt
```

```python
improvements = [
    '**Specificity of Output:** The new prompt clearly defines the desired output format (2-bullet summary, 3 MCQs with specific correct/distractor ratio), unlike the vague \"Make a quiz\".',
    '**Target Audience and Tone:** It explicitly states the target audience (7th graders / 11-13 y/o) and the required tone (friendly, simple vocabulary), ensuring content is tailored appropriately.',
    '**Utility and Format Constraint:** By requesting \"Google Slides-ready output,\" the prompt ensures the generated content is immediately usable and structured for a specific platform, adding practical value.',
]
improvements
```

## Exercise 4: Match Prompt to Purpose
Pick a style (Exploratory, Structured, Conversational, Functional) to build a customer-support chatbot prompt.

### Tasks
- Write the full prompt in the chosen style.
- Identify 2 features that prove it matches the style.
- Justify why this style fits better than the others.

```python
chosen_style = 'Conversational'
prompt_template = """You are a friendly and helpful customer support chatbot for "AquaFlow Smart Sprinklers". Your goal is to assist users with common issues and questions about their product.

When a user describes a problem or asks a question, first acknowledge their situation with empathy. Then, ask clarifying questions to understand the issue fully. Finally, provide clear, step-by-step solutions or direct them to relevant resources.

If you don't know the answer, politely state that you're unable to help and offer to escalate to a human agent.

Maintain a positive and supportive tone throughout the interaction."""
key_style_features = [
    'Empathy and rapport-building: The prompt explicitly asks the chatbot to acknowledge situations with empathy and maintain a positive, supportive tone.',
    'Interactive problem-solving: It guides the chatbot to ask clarifying questions and provide step-by-step solutions, fostering a natural dialogue.',
]
justification = """The 'Conversational' style is best for a customer support chatbot because it allows for natural, empathetic interactions, helping users feel heard and understood. Unlike 'Functional' (which might be too rigid), 'Structured' (too formal), or 'Exploratory' (too open-ended), a conversational approach can handle nuanced problems, build rapport, and guide users through troubleshooting with a friendly demeanor, leading to better customer satisfaction."""
print(chosen_style)
print(prompt_template)
print(key_style_features)
print(justification)
```

## Exercise 6: Hallucination Spotting and Mitigation
**Scenario**: Model added an unsupported claim in a climate/marine article summary.

### Tasks
1) Prompt the model to avoid hallucinations using only provided text.
2) Rewrite with verification language.
3) List 2 mitigation strategies.
4) Name 2 domains where hallucinations are risky and why.

```python
no_hallucination_prompt = """Summarize the following climate/marine article. Only use information explicitly stated in the provided text. Do not introduce any outside knowledge or make assumptions.\n\nArticle: [INSERT ARTICLE TEXT HERE]"""
no_hallucination_prompt
```

```python
verification_prompt = """Analyze the following climate/marine article and provide a summary. For every factual statement in your summary, indicate in parentheses the paragraph number from the article where the information can be verified. If a piece of information cannot be directly supported by the text, explicitly state 'Information not found in article'. Do not infer or extrapolate.\n\nArticle: [INSERT ARTICLE TEXT HERE]"""
verification_prompt
```

```python
mitigation_strategies = [
    '**Grounding:** Explicitly instruct the model to only use provided source text and to avoid external knowledge.',
    '**Verification/Attribution:** Ask the model to cite sources or indicate where information can be found within the provided text.',
]
risky_domains = {
    'domain_1': {
        'name': 'Medical/Healthcare',
        'why': 'Hallucinations in medical advice or diagnoses can lead to serious health consequences, misdiagnosis, or inappropriate treatment, directly impacting patient safety and well-being.',
    },
    'domain_2': {
        'name': 'Legal/Financial Advice',
        'why': 'Inaccurate or fabricated information in legal or financial contexts can result in significant legal liabilities, financial losses, incorrect compliance, and severe reputational damage.',
    },
}
print(mitigation_strategies)
print(risky_domains)
```
