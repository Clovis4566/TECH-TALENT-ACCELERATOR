## рџЊџ Exercise 2: Vectorization with Sentence Transformers

In this exercise, we will transform our textual data (news titles) into numerical representations known as embeddings using the Sentence Transformers library. This is a crucial step for enabling machines to process and understand text in downstream tasks.

### 1. Prepare the Data for Embedding Generation

We will use the `pdf_subset` created earlier. The `InputExample` class helps format data for sentence transformers.

```python
# We already have pdf_subset from the previous exercise.
```

### 2. Create a Helper Function

This function converts each record (news title) from the DataFrame into an `InputExample` object, which is the required format for the embedding process.

```python
def example_create_fn(doc1: pd.Series) -> InputExample:
    """
    Helper function that outputs a sentence_transformer guid, label, and text.
    """
    return InputExample(guid=doc1["id"], texts=[doc1["title"]])
```

### 3. Apply the Helper Function to the Subset

We'll apply this function to our `pdf_subset` to generate a list of `InputExample` objects, preparing the data for embedding.

```python
faiss_train_examples = pdf_subset.apply(lambda x: example_create_fn(x["title"]), axis=1).tolist()
display(faiss_train_examples[:5]) # Display first 5 examples
```

### 4. Initialize the Embedding Model

We will use the `all-MiniLM-L6-v2` pre-trained model from Sentence Transformers, which is efficient and provides high-quality embeddings.

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

### 5. Extract the Titles and Convert to a List of Strings

Prepare the raw text titles from our subset into a list, as this is the input format expected by the model for encoding.

```python
titles_list = pdf_subset["title"].tolist()
```

### 6. Generate Embeddings for the Titles

Now, we will use the initialized model to generate a dense vector embedding for each title in our list.

```python
faiss_title_embedding = model.encode(titles_list, show_progress_bar=True)
```

### 7. Check Embedding Dimensions

To verify that the embeddings were generated correctly, we'll check the total number of embeddings and the dimensionality of each embedding vector.

```python
print(f"Number of embeddings: {len(faiss_title_embedding)}")
print(f"Dimension of each embedding: {len(faiss_title_embedding[0])}")
```
