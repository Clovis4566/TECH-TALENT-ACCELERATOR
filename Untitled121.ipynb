## Step 3: Create Custom MCP Server
We define a server with two tools: `summarize_text` and `generate_changelog`.

```python
from pathlib import Path
import textwrap

server_path = Path("/content/custom_workspace_mcp.py")
server_path.write_text(textwrap.dedent("""
from fastmcp import FastMCP
from typing import List

mcp = FastMCP(name="workspace_utils")

@mcp.tool
def summarize_text(text: str) -> str:
    \"\"\"Analyzes text and returns stats and a brief summary.\"\"\"
    lines = text.splitlines()
    words = text.split()
    summary = text[:100] + "..." if len(text) > 100 else text
    return f"Stats: {len(lines)} lines, {len(words)} words.\\nSummary: {summary}"

@mcp.tool
def generate_changelog(changes: List[str]) -> str:
    \"\"\"Generates a formatted Markdown changelog from a list of strings.\"\"\"
    items = "\\n".join([f"- {c}" for c in changes])
    return f"### Workspace Changelog\\n\\n{items}"

if __name__ == \"__main__\":
    mcp.run(transport=\"stdio\")
"""), encoding="utf-8")

print(f"Wrote custom MCP server to {server_path}")
```

## Step 4 & 5: Connect Servers and Build Agent
We orchestrate the Filesystem server and our custom server using `MultiServerMCPClient` and build the Gemini agent.

```python
import asyncio
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.prebuilt import create_react_agent

async def setup_agent():
    # 1. Define connections
    mcp_connections = {
        "filesystem": {
            "transport": "stdio",
            "command": "npx",
            "args": ["-y", "@modelcontextprotocol/server-filesystem", WORKDIR],
        },
        "custom_utils": {
            "transport": "stdio",
            "command": "python",
            "args": [str(server_path)],
        }
    }

    # 2. Initialize Client and Get Tools
    client = MultiServerMCPClient(mcp_connections, tool_name_prefix=True)
    tools = await client.get_tools()

    # 3. Initialize Gemini
    llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash")

    # 4. Create Agent
    agent = create_react_agent(llm, tools)
    return agent

# Create the agent instance
workspace_agent = asyncio.get_event_loop().run_until_complete(setup_agent())
print("Agent is ready with MCP tools!")
```

## Step 6: Demonstration
Let's ask the agent to perform the multi-step task.

```python
async def run_demo():
    query = """
    Please perform the following steps:
    1. Create a file named 'notes.txt' in the workspace with content about Artificial Intelligence.
    2. Read the content of 'notes.txt'.
    3. Use the custom summarization tool to analyze that content.
    4. Generate a changelog entry for this action and save the summary and changelog into a new file called 'report.md'.
    """

    config = {"configurable": {"thread_id": "demo_1"}}
    async for event in workspace_agent.astream({"messages": [("user", query)]}, config):
        for node, value in event.items():
            if "messages" in value:
                value["messages"][-1].pretty_print()

asyncio.get_event_loop().run_until_complete(run_demo())
```
