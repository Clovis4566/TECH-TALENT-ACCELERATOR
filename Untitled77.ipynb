```python
import os, math, re, random
from glob import glob
from pathlib import Path
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

np.random.seed(42); tf.random.set_seed(42)

# Paths mis à jour pour correspondre au téléchargement
DATA_ROOT = Path("data/cats_dogs")
train_dir = DATA_ROOT / "train"

# Suppression des fichiers corrompus potentiels (fréquent dans ce dataset)
for folder in ["cat", "dog"]:
    folder_path = train_dir / folder
    for img_path in folder_path.glob("*.jpg"):
        if os.path.getsize(img_path) == 0:
            os.remove(img_path)

IMG_HEIGHT, IMG_WIDTH = 128, 128 # Réduit pour la vitesse
batch_size = 32
seed = 1337

def build_df_from_folder(folder: Path):
    files = list(folder.glob('**/*.jpg'))
    rows = [{"filepath": str(f), "label": f.parent.name} for f in files]
    return pd.DataFrame(rows)

df_train_full = build_df_from_folder(train_dir)

# Split
from sklearn.model_selection import train_test_split
df_tr, df_val = train_test_split(df_train_full, test_size=0.2, stratify=df_train_full["label"], random_state=seed)

# Générateurs
train_gen = ImageDataGenerator(rescale=1./255, rotation_range=20, horizontal_flip=True)
val_gen = ImageDataGenerator(rescale=1./255)

train_flow = train_gen.flow_from_dataframe(df_tr, x_col="filepath", y_col="label", target_size=(IMG_HEIGHT, IMG_WIDTH), class_mode="binary", batch_size=batch_size)
val_flow = val_gen.flow_from_dataframe(df_val, x_col="filepath", y_col="label", target_size=(IMG_HEIGHT, IMG_WIDTH), class_mode="binary", batch_size=batch_size)

print(f"Images trouvées : {len(df_train_full)}")
```

### 2. Inspection des données

**Rapport rapide :**
- **Nombre total d'images :** 24,998
- **Répartition :** Environ 12,500 images par classe (équilibré).
- **Variabilité :** Les images contiennent des variations de luminosité, de pose (couché, debout), et de zoom.

**Indices visuels :** Pour distinguer les deux, le modèle apprendra des textures (poils), la forme des oreilles (pointues vs tombantes) et la structure du museau.

```python
def plot_images(gen_flow):
    x, y = next(gen_flow)
    plt.figure(figsize=(12, 12))
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(x[i])
        label = "Chien" if y[i] == 1 else "Chat"
        plt.title(label)
        plt.axis("off")
    plt.show()

plot_images(train_flow)
```

### 3 & 4. Architecture et Optimisation

- **Modèle :** CNN avec 3 blocs (Conv2D + MaxPooling).
- **Régularisation :** Dropout à 0.5 pour éviter que le modèle ne retienne les images par cœur.
- **Optimiseur :** Adam (très efficace pour débuter).
- **Fonction de perte :** Binary Crossentropy (car nous avons 2 choix possibles).

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    # Premier bloc : cherche des formes simples
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    layers.MaxPooling2D((2, 2)),

    # Deuxième bloc : formes plus complexes
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Troisième bloc
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Conversion en vecteur pour la classification finale
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5), # Anti-triche (overfitting)
    layers.Dense(1, activation='sigmoid') # Résultat final : 0 ou 1
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

### 5. Entraînement du modèle

Nous entraînons le modèle avec un rappel (**callback**) d'arrêt précoce pour éviter le surapprentissage. Si la perte de validation ne s'améliore pas pendant 3 époques, l'entraînement s'arrête.

```python
from tensorflow.keras.callbacks import EarlyStopping

# Arrêt précoce pour gagner du temps et éviter l'overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(
    train_flow,
    validation_data=val_flow,
    epochs=10,
    callbacks=[early_stop]
)
```

### 5 (suite). Visualisation des performances
Nous affichons les graphiques pour vérifier si le modèle a bien généralisé ou s'il a surappris.

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Précision (Accuracy)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Perte (Loss)')
plt.legend()

plt.show()
```

### 6. Évaluation sur les données de validation
Nous calculons la matrice de confusion pour voir précisément quelles erreurs le modèle commet.

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# 1. Prédire sur les données de validation
val_flow.reset()
y_pred_probs = model.predict(val_flow)
y_pred = (y_pred_probs > 0.5).astype(int)
y_true = val_flow.classes

# 2. Afficher le rapport de classification
print("Rapport de Classification :")
print(classification_report(y_true, y_pred, target_names=['Chat', 'Chien']))

# 3. Matrice de confusion
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Chat', 'Chien'], yticklabels=['Chat', 'Chien'])
plt.xlabel('Prédiction')
plt.ylabel('Réalité')
plt.title('Matrice de Confusion')
plt.show()
```
