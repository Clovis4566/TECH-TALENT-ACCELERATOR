```python
from datasets import load_dataset
from collections import Counter

print("Téléchargement sécurisé du dataset tweet_eval (sentiment)...")
# Utilisation du chemin mis à jour et stable pour éviter l'erreur HfUriError
dataset = load_dataset("cardiffnlp/tweet_eval", "sentiment")
print("\n--- Structure des splits ---")
print(dataset)

# Mappage officiel des labels pour tweet_eval sentiment
id2label = {0: "negative", 1: "neutral", 2: "positive"}

print("\n--- Distribution des classes (Train) ---")
train_labels = dataset["train"]["label"]
distribution = Counter(train_labels)
for k, v in distribution.items():
    print(f"Label {k} ({id2label[k]}) : {v} tweets")

# Sauvegarde de deux exemples par classe pour la suite
saved_examples = { "negative": [], "neutral": [], "positive": [] }

for item in dataset["train"]:
    label_str = id2label[item["label"]]
    if len(saved_examples[label_str]) < 2:
        saved_examples[label_str].append(item["text"])
    if all(len(v) == 2 for v in saved_examples.values()):
        break

print("\n--- Exemples sauvegardés pour plus tard ---")
for label_str, tweets in saved_examples.items():
    print(f"\n[{label_str.upper()}]")
    for i, t in enumerate(tweets, 1):
        print(f"  {i}. {t}")
```

```python
from transformers import AutoTokenizer

print("Initialisation du Tokenizer DistilBERT...")
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def preprocess_function(examples):
    # Encodage complet avec troncature et padding automatique à 128 tokens
    return tokenizer(
        examples["text"],
        max_length=128,
        padding="max_length",
        truncation=True
    )

print("Tokenisation des différents splits en cours...")
# On mappe la fonction en activant le traitement par lots pour accélérer l'exécution
tokenized_dataset = dataset.map(preprocess_function, batched=True)

# Application du format PyTorch pour le Hugging Face Trainer API
tokenized_dataset.set_format("torch", columns=["input_ids", "attention_mask", "label"])

print("\n--- Tokenisation terminée ---")
print("Clés disponibles dans l'échantillon d'entraînement :", tokenized_dataset["train"].column_names)
```

```python
import numpy as np
import evaluate
import torch
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
from transformers import DataCollatorWithPadding

print("Chargement de DistilBERT pour la classification (3 classes)...")
model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=3
)

# 1. On réinitialise le format brut pour contourner le bug torchvision
tokenized_dataset.reset_format()

# Chargement des métriques
accuracy_metric = evaluate.load("accuracy")
f1_metric = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    acc = accuracy_metric.compute(predictions=predictions, references=labels)["accuracy"]
    f1_macro = f1_metric.compute(predictions=predictions, references=labels, average="macro")["f1"]
    return {"accuracy": acc, "f1_macro": f1_macro}

# Configuration des arguments
training_args = TrainingArguments(
    output_dir="./results_distilbert_sentiment",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",
    logging_steps=20,
    report_to="none"
)

# Sélection des sous-ensembles
train_sub = tokenized_dataset["train"].shuffle(seed=42).select(range(4000))
val_sub = tokenized_dataset["validation"].shuffle(seed=42).select(range(500))

# 2. Utilisation d'un DataCollator qui formate les données proprement en tenseurs à la volée
data_collator = DataCollatorWithPadding(tokenizer=tokenizer, return_tensors="pt")

# Initialisation du Trainer mis à jour
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_sub,
    eval_dataset=val_sub,
    compute_metrics=compute_metrics,
    data_collator=data_collator # Convertit en tenseurs PyTorch en toute sécurité
)

print("\nDébut du Fine-Tuning de DistilBERT...")
trainer.train()

print("\nSauvegarde du meilleur modèle...")
trainer.save_model("./best_sentiment_model")
tokenizer.save_pretrained("./best_sentiment_model")
print("Modèle enregistré avec succès !")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoModel, AutoModelForSequenceClassification

# ==============================================================================
# 4. ÉVALUATION & HISTOGRAMME DE CALIBRATION
# ==============================================================================
print("Calcul des scores de confiance sur l'ensemble de test...")
test_sub = tokenized_dataset["test"].shuffle(seed=42).select(range(500))

# Extraction des prédictions brutes
predictions_output = trainer.predict(test_sub)
logits = predictions_output.predictions

# Conversion des logits en probabilités via Softmax (PyTorch)
probabilities = torch.nn.functional.softmax(torch.tensor(logits), dim=-1).numpy()
# On récupère la confiance de la classe prédite
confidences = np.max(probabilities, axis=1)

# Tracé de l'histogramme de calibration
plt.figure(figsize=(8, 5))
sns.histplot(confidences, bins=10, kde=True, color="skyblue", edgecolor="black")
plt.title("Distribution des scores de confiance (Calibration)")
plt.xlabel("Confiance (Score Softmax maximum)")
plt.ylabel("Nombre de Tweets")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("\n--- Analyse de la calibration ---")
print(f"Confiance moyenne du modèle : {np.mean(confidences):.3f}")
print("Commentaire : Si l'histogramme est très écrasé vers 0.9-1.0 alors que l'accuracy globale")
print("est plus basse, le modèle souffre de sur-confiance (over-confidence), classique avec BERT.")

# ==============================================================================
# 5. INSPECTION DES CARTES D'ATTENTION
# ==============================================================================
print("\nExtraction des poids de la couche d'attention pour un exemple...")

# Choix d'une phrase textuelle (Exemple négatif sauvegardé plus tôt)
sample_tweet = "So disappointed in wwe summerslam! I want to see john cena wins"

# On charge le modèle de base pour pouvoir extraire les sorties d'attention (output_attentions=True)
model_base = AutoModel.from_pretrained("./best_sentiment_model", output_attentions=True)
model_base.eval()

inputs = tokenizer(sample_tweet, return_tensors="pt")
with torch.no_grad():
    outputs = model_base(**inputs)

# outputs.attentions contient les poids d'attention pour chaque couche
# Structure : (num_layers, batch_size, num_heads, seq_len, seq_len)
last_layer_attention = outputs.attentions[-1][0]  # Dernière couche, premier exemple du batch

# Moyenne des poids d'attention sur l'ensemble des têtes (Heads) du Transformer
mean_attention = torch.mean(last_layer_attention, dim=0).numpy()

# Récupération des tokens réels pour les étiquettes du graphique
tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

# Extraction de la ligne d'attention dirigée depuis le token spécial [CLS] vers tous les autres
cls_attention = mean_attention[0, :]

# Visualisation sous forme de Heatmap
plt.figure(figsize=(12, 2))
sns.heatmap([cls_attention], annot=True, fmt=".3f", xticklabels=tokens, yticklabels=["[CLS]"], cmap="YlOrRd", cbar=False)
plt.title("Importance des tokens mesurée par l'attention du jeton [CLS] (Dernière Couche)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np

# Re-chargement du modèle de classification entraîné pour l'inférence finale
final_model = AutoModelForSequenceClassification.from_pretrained("./best_sentiment_model")
final_model.eval()

# Chargement du modèle de base pour extraire les cartes d'attention
attention_extractor = AutoModel.from_pretrained("./best_sentiment_model", output_attentions=True)
attention_extractor.eval()

# Mappage des étiquettes de notre dataset
labels_map = {0: "Negative", 1: "Neutral", 2: "Positive"}

def analyze_text(text: str):
    # 1. Préparation et tokenisation de la chaîne de texte brute
    inputs = tokenizer(text, return_tensors="pt", max_length=128, padding=True, truncation=True)
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

    with torch.no_grad():
        # 2. Calcul des probabilités (Prediction & Confidence)
        logits = final_model(**inputs).logits
        probs = torch.nn.functional.softmax(logits, dim=-1).numpy()[0]
        predicted_class = int(np.argmax(probs))

        # 3. Extraction de l'importance des mots (Attention)
        outputs_att = attention_extractor(**inputs)
        last_layer_att = outputs_att.attentions[-1][0] # Dernière couche
        mean_attention = torch.mean(last_layer_att, dim=0).numpy() # Moyenne des têtes
        cls_attention_weights = mean_attention[0, :] # Ligne du jeton spécial [CLS]

    # Associer chaque token à sa note d'attention (en ignorant les jetons de structure [CLS], [SEP])
    token_importance = []
    for idx, (token, weight) in enumerate(zip(tokens, cls_attention_weights)):
        if token not in ["[CLS]", "[SEP]", "[PAD]"]:
            token_importance.append((token, float(weight)))

    # Extraire les 3 jetons ayant reçu le plus d'attention de la part du modèle
    token_importance.sort(key=lambda x: x[1], reverse=True)
    highlighted_tokens = token_importance[:3]

    return {
        "label": labels_map[predicted_class],
        "confidence": float(probs[predicted_class]),
        "highlighted_tokens": highlighted_tokens
    }

# --- TEST DE PRODUCTION EN DIRECT ---
print("Test de la fonction de production analyze_text() :")
tweet_test = "This customer service update is terrible, absolute garbage."
result = analyze_text(tweet_test)

print(f"\nTexte analysé : '{tweet_test}'")
print(f"Résultat      : {result}")
```
