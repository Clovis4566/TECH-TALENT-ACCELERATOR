```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow import keras
import matplotlib.pyplot as plt
```

### Load the IMDB Dataset

We'll load the IMDB dataset, keeping only the 10,000 most frequent words. The data comes pre-split into training and testing sets.

```python
num_words = 10000  # keep the 10k most frequent tokens

(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=num_words)

print(f"Training data shape: {train_data.shape}")
print(f"Training labels shape: {train_labels.shape}")
print(f"Test data shape: {test_data.shape}")
print(f"Test labels shape: {test_labels.shape}")

# Display an example review and its label
print(f"\nExample review (first 10 elements): {train_data[0][:10]}")
print(f"Example label: {train_labels[0]}")
```

### Preprocessing the Data: One-Hot Encoding

The reviews are currently sequences of integers. We need to convert them into binary matrices using one-hot encoding. This means each review will become a vector where indices corresponding to words present in the review are marked with a '1', and others with a '0'.

```python
def vectorize_sequences(sequences, dimension=10000):
    # Create an all-zero matrix of shape (len(sequences), dimension)
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        # Set specific indices of results[i] to 1s
        results[i, sequence] = 1.
    return results

# Vectorize training and test data
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)

# Vectorize labels (convert to float32)
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')

print(f"Vectorized x_train shape: {x_train.shape}")
print(f"Vectorized x_test shape: {x_test.shape}")
print(f"Vectorized y_train shape: {y_train.shape}")
print(f"Vectorized y_test shape: {y_test.shape}")

# Display an example of a vectorized review
print(f"\nExample vectorized review (first 10 elements): {x_train[0][:10]}")
print(f"Example vectorized review (sum of elements, should be number of unique words): {np.sum(x_train[0])}")
```

### Splitting Data into Training, Validation, and Test Sets

Although `imdb.load_data` already provides training and test sets, it's good practice to create a validation set from the training data to monitor model performance during training and detect overfitting early. We'll use the first 10,000 samples of the training data for validation.

```python
x_val = x_train[:10000]
y_val = y_train[:10000]

partial_x_train = x_train[10000:]
partial_y_train = y_train[10000:]

print(f"Partial training data shape: {partial_x_train.shape}")
print(f"Partial training labels shape: {partial_y_train.shape}")
print(f"Validation data shape: {x_val.shape}")
print(f"Validation labels shape: {y_val.shape}")
```

### Building the Model

We'll construct a simple feedforward neural network. Given our input data are vectors and labels are scalars, a stack of `Dense` layers with `relu` activations is suitable. For binary classification with a probability output, `binary_crossentropy` is the appropriate loss function.

```python
from tensorflow.keras import models
from tensorflow.keras import layers

# Define the model architecture
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

### Training the Model

Now we'll train our model. We'll use a `batch_size` of 512 and train for 20 `epochs`. The `validation_data` will be used to monitor the model's performance on unseen data during training, which helps us detect overfitting.

```python
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=20,
                    batch_size=512,
                    validation_data=(x_val, y_val))

history_dict = history.history
print(history_dict.keys())
```

### Evaluating the Model: Visualizing Training and Validation Metrics

To detect overfitting, we'll plot the training and validation loss, as well as training and validation accuracy over the epochs. This allows us to see how well the model generalizes to unseen data.

```python
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']

epochs = range(1, len(loss_values) + 1)

plt.figure(figsize=(10, 6))
plt.plot(epochs, loss_values, 'bo', label='Training loss')  # 'bo' for blue dot
plt.plot(epochs, val_loss_values, 'b', label='Validation loss') # 'b' for solid blue line
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']

plt.figure(figsize=(10, 6))
plt.plot(epochs, acc_values, 'bo', label='Training accuracy')
plt.plot(epochs, val_acc_values, 'b', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

### Retraining the Model with Optimal Epochs and Final Evaluation

From the plots, we can observe the point where the validation loss starts to increase (or validation accuracy starts to decrease), indicating overfitting. We'll retrain the model for an optimal number of epochs before it starts to overfit, and then evaluate its performance on the completely unseen test set.

```python
# Retrain a new model from scratch with the optimal number of epochs
# (e.g., if plots suggest 4 epochs is optimal)
# Re-initialize the model to train from scratch
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train for fewer epochs to avoid overfitting
# You might need to adjust the number of epochs based on the plots above
optimal_epochs = 4 # Example: adjust this based on visual inspection of plots
model.fit(partial_x_train,
          partial_y_train,
          epochs=optimal_epochs,
          batch_size=512,
          validation_data=(x_val, y_val))

# Evaluate the model on the test data
results = model.evaluate(x_test, y_test)

print(f"\nFinal Test Loss: {results[0]:.4f}")
print(f"Final Test Accuracy: {results[1]:.4f}")
```
