# Precision Prompting: Resource Access Security

## Scenario
You work on the Learning & Communications Team at a global tech company. Your task is to turn a dense internal policy update into a concise microlearning snippet for an internal newsletter.

### Input Text:
> "Employees must ensure that all remote access to internal systems is established via the approved secure VPN. Under no circumstances should unsecured connections or personal devices lacking endpoint protection be used to access proprietary data or sensitive communications."

## Step 1 & 2: Craft and Evaluate a Prompt

**Task:** Write a prompt that rewrites the input in a friendly tone, uses bullet points, paraphrases (no quotes), and stays under 75 words.

```python
# Define the input text
policy_text = """Employees must ensure that all remote access to internal systems is established via the approved secure VPN. Under no circumstances should unsecured connections or personal devices lacking endpoint protection be used to access proprietary data or sensitive communications."""

# Step 1: Craft a Prompt That Controls Output Tone, Format, and Length
prompt_step_1 = f"""
Rewrite the following policy text in a friendly and clear tone for an internal newsletter.

Constraints:
- Use bullet points to organize the information.
- Paraphrase the content; do not use direct quotes.
- Keep the total length under 75 words.
- Do not add new information or technologies.

Input Text: {policy_text}
"""

print("Step 1 Prompt:")
print(prompt_step_1)
```

## Step 3 & 4: Mitigation and Paraphrasing Deep Dive

**Task:** Refine the prompt to prevent hallucinations and create a version specifically for junior interns using plain language.

```python
# Step 4: Junior Intern Audience Prompt (Plain Language)
prompt_interns = f"""
Paraphrase the policy below for a junior intern audience using very simple, supportive language.

Constraints:
- Use no more than 4 bullet points.
- Avoid all corporate or legal jargon.
- Keep the tone helpful and informative.
- Limit your response to the facts provided in the input.

Input Text: {policy_text}
"""

print("Step 4 Prompt:")
print(prompt_interns)
```

## Step 5: Quote Extraction

**Task:** Extract one direct quote and discuss the use cases for quoting vs. paraphrasing.

```python
# Step 5: Quote Extraction Prompt
prompt_quote = f"""
Extract one single direct quote from the following text that most effectively captures the core security requirement.

Input Text: {policy_text}
"""

print("Step 5 Prompt:")
print(prompt_quote)
```

### Step 5 Discussion Questions

**1. In what kind of internal communication would quoting be more appropriate than paraphrasing?**
Quoting is most appropriate in formal legal notices, official policy declarations, or disciplinary communications where the exact wording of a rule is critical to avoid ambiguity. It is also useful when the original author has a high level of authority and their specific voice adds weight to the message.

**2. When might quoting pose a risk?**
Quoting poses a risk when the original text is overly dense, jargon-heavy, or formal. In these cases, the audience may fail to grasp the practical application of the policy. It can also appear cold or impersonal in a culture that values a friendly, conversational tone, and it fails to simplify complex ideas for non-technical staff.
