```python
# ==============================================================================
# PROJET SENTIMENT ASSISTANT AVEC HUGGING FACE PIPELINE (OPTIMISÉ COLAB)
# ==============================================================================

# 1. Installation propre des packages requis
print("[DEBUG] 1/6 - Installation des bibliothèques...")
!pip install -q transformers datasets accelerate

import torch
import platform
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

# 2. Vérification de la configuration Matériel
print("\n[DEBUG] 2/6 - Vérification du GPU...")
device = 0 if torch.cuda.is_available() else -1
if device == 0:
    print(f"✅ [OK] GPU détecté : {torch.cuda.get_device_name(0)}. BERT tournera à pleine vitesse !")
else:
    print("⚠️ [ATTENTION] Aucun GPU détecté. Allez dans Exécution > Modifier le type d'exécution et activez le T4 GPU.")

# 3. Chargement instantané du modèle BERT Fine-Tuned (ou checkpoint optimisé)
print("\n[DEBUG] 3/6 - Chargement du modèle BERT et du Tokenizer...")
model_name = "distilbert-base-uncased-finetuned-sst-2-english" # Version optimisée et plus rapide de BERT

try:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    print(f"✅ [OK] Modèle '{model_name}' chargé avec succès.")
except Exception as e:
    print(f"❌ [ERREUR] Échec du téléchargement du modèle : {e}")

# 4. Création de la pipeline d'analyse de sentiment
print("\n[DEBUG] 4/6 - Initialisation de la pipeline d'inférence...")
try:
    sentiment_analyzer = pipeline(
        "sentiment-analysis",
        model=model,
        tokenizer=tokenizer,
        device=device
    )
    print("✅ [OK] Pipeline prête à l'emploi.")
except Exception as e:
    print(f"❌ [ERREUR] Échec de la création de la pipeline : {e}")

# 5. Définition de l'assistant d'inférence demandé par le projet
print("\n[DEBUG] 5/6 - Création de la fonction predict_sentiment...")
def predict_sentiment(text: str):
    try:
        # La pipeline s'occupe de l'encodage, de la troncature (max_length) et du Softmax
        result = sentiment_analyzer(text, truncation=True, max_length=256)[0]

        # Formatage pour correspondre aux attentes du projet (Positive/Negative)
        label = "Positive" if result['label'] == "POSITIVE" else "Negative"
        confidence = result['score']

        return label, float(confidence)
    except Exception as e:
        print(f"❌ [ERREUR INFÉRENCE] Analyse impossible : {e}")
        return "Error", 0.0

# 6. Exécution du test sur un cas réel de support client
print("\n[DEBUG] 6/6 - Exécution du test final...")
custom_sentence = "The onboarding emails were confusing, but the agent fixed everything politely."
label, confidence = predict_sentiment(custom_sentence)

print("-" * 60)
print(f"Phrase Client : '{custom_sentence}'")
print(f"Résultat      : {label}")
print(f"Confiance     : {confidence * 100:.2f}%")
print("-" * 60)
```
