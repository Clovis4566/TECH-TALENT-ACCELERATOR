```python
import platform
import numpy as np
import tensorflow as tf

# ==============================================================================
# 1. VÉRIFICATION DU MATÉRIEL
# ==============================================================================
print("Python version      :", platform.python_version())
print("TensorFlow version  :", tf.__version__)
print("GPU devices detected:", tf.config.list_physical_devices('GPU'))

# ==============================================================================
# 2. CHARGEMENT ET PRÉPARATION DU JEU DE DONNÉES IMDB
# ==============================================================================
print("\nChargement du jeu de données IMDb via Keras...")
MAX_WORDS = 10000
MAX_LENGTH = 256
BATCH_SIZE = 16

# Chargement natif des données IMDb sous forme d'index de mots
(x_train_raw, y_train), (x_test_raw, y_test) = tf.keras.datasets.imdb.load_data(num_words=MAX_WORDS)

# Extraction du dictionnaire de mots pour décoder et normaliser notre texte
word_index = tf.keras.datasets.imdb.get_word_index()
reverse_word_index = {value + 3: key for key, value in word_index.items()}
reverse_word_index[0] = "<PAD>"
reverse_word_index[1] = "<START>"
reverse_word_index[2] = "<UNK>"
reverse_word_index[3] = "<UNUSED>"

def decode_review(text_ids):
    return " ".join([reverse_word_index.get(i, "?") for i in text_ids])

# Re-constitution du texte brut pour l'entraînement et le test
print("Préparation des textes pour le TextVectorization...")
train_texts = [decode_review(seq) for seq in x_train_raw[:2000]]
test_texts = [decode_review(seq) for seq in x_test_raw[:500]]

Y_train = y_train[:2000]
Y_test = y_test[:500]

# 🧠 Remplacement du Tokenizer Hugging Face par le TextVectorization natif de Keras
# C'est la solution la plus stable sous TensorFlow pour découper le texte en jetons (tokens)
vectorize_layer = tf.keras.layers.TextVectorization(
    max_tokens=MAX_WORDS,
    output_mode='int',
    output_sequence_length=MAX_LENGTH
)

# Adaptation du tokenizer au vocabulaire d'IMDb
vectorize_layer.adapt(train_texts)
print("Tokenizer Keras adapté avec succès au vocabulaire !")

# Transformation des textes en tenseurs d'identifiants numériques
X_train_seq = vectorize_layer(np.array(train_texts))
X_test_seq = vectorize_layer(np.array(test_texts))

# Création du pipeline tf.data.Dataset (shuffle, batch, prefetch) exigé par le projet
train_ds = tf.data.Dataset.from_tensor_slices((X_train_seq, Y_train)).shuffle(2000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_ds = tf.data.Dataset.from_tensor_slices((X_test_seq, Y_test)).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# ==============================================================================
# 3. INITIALISER LE MODÈLE DE CLASSIFICATION (KERAS PUR)
# ==============================================================================
print("\nInitialisation du modèle de classification...")

# Structure d'entrée séquentielle
input_seq = tf.keras.layers.Input(shape=(MAX_LENGTH,), dtype=tf.int32, name="input_sequence")

# Architecture d'analyse de sentiments
embedding_layer = tf.keras.layers.Embedding(input_dim=MAX_WORDS, output_dim=128, input_length=MAX_LENGTH)(input_seq)
global_pool = tf.keras.layers.GlobalAveragePooling1D()(embedding_layer)
dropout = tf.keras.layers.Dropout(0.2)(global_pool)
output_logits = tf.keras.layers.Dense(2, name="logits")(dropout)

model = tf.keras.Model(inputs=input_seq, outputs=output_logits)

# Compilation standard (Adam, CrossEntropy avec logits pour la stabilité)
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-5, epsilon=1e-8)
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metrics = [tf.keras.metrics.SparseCategoricalAccuracy(name="accuracy")]

model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrics)
model.summary()

# ==============================================================================
# 4. FORMER ET SUPERVISER
# ==============================================================================
EPOCHS = 2
print(f"\nDébut de l'entraînement pour {EPOCHS} époques via model.fit()...")
history = model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=EPOCHS
)

# ==============================================================================
# 5. ÉVALUER SUR L'ENSEMBLE DE TEST
# ==============================================================================
print("\nÉvaluation finale sur l'ensemble de test...")
eval_metrics = model.evaluate(test_ds)
print(f"Loss globale de test : {eval_metrics[0]:.4f}")
print(f"Précision (Accuracy) globale de test : {eval_metrics[1]:.4f}")

# ==============================================================================
# 6. CRÉER UN ASSISTANT D'INFÉRENCE RÉUTILISABLE
# ==============================================================================
def predict_sentiment(text: str):
    # Transformation de la chaîne de texte brute via notre tokenizer Keras
    input_tensor = vectorize_layer(np.array([text]))

    # Inférence
    logits = model(input_tensor)
    probs = tf.nn.softmax(logits, axis=-1).numpy()[0]

    # Extraction de la prédiction (0 = Négatif, 1 = Positif)
    label_id = tf.argmax(probs).numpy()
    label = "Positive" if label_id == 1 else "Negative"

    return label, float(probs[label_id])

print("\nTest de l'assistant d'inférence en direct :")
custom_sentence = "The onboarding emails were confusing, but the agent fixed everything politely."
label, confidence = predict_sentiment(custom_sentence)
print(f"Phrase  : {custom_sentence}")
print(f"Prediction: {label} (confidence={confidence:.3f})")
```
