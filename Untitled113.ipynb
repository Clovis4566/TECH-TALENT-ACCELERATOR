# 2. Load the dataset
Nous utilisons le dataset `databricks-dolly-15k` et extrayons la colonne `context`.

```python
from langchain_community.document_loaders import HuggingFaceDatasetLoader

dataset_name = "databricks/databricks-dolly-15k"
page_content_column = "context"

loader = HuggingFaceDatasetLoader(dataset_name, page_content_column)
data = loader.load()
print(f"Loaded {len(data)} documents.")
print("Sample content:", data[0].page_content[:200])
```

# 3. Split the documents
Le texte est divisГ© en morceaux de 1000 caractГЁres avec un chevauchement pour garder le contexte.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
docs = text_splitter.split_documents(data)
print(f"Number of chunks: {len(docs)}")
```

# 4. Embed the text
On transforme les mots en vecteurs mathГ©matiques comprГ©hensibles par la machine.

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

modelPath = "sentence-transformers/all-MiniLM-l6-v2"
model_kwargs = {'device':'cpu'}
encode_kwargs = {'normalize_embeddings': False}

embeddings = HuggingFaceEmbeddings(
    model_name=modelPath,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

# 5. Create a vector store
On utilise FAISS pour indexer nos vecteurs afin de pouvoir chercher rapidement.

```python
from langchain_community.vectorstores import FAISS

# Note: This might take a few minutes on a large dataset
db = FAISS.from_documents(docs[:500], embeddings) # Limitation Г  500 pour la rapiditГ© de dГ©mo
print("Vector store created.")
```

# 6. Prepare the LLM model
Configuration du pipeline de Question-Answering avec TinyBERT.

```python
from transformers import AutoTokenizer, pipeline
from langchain_community.llms import HuggingFacePipeline

model_name = "Intel/dynamic_tinybert"
tokenizer = AutoTokenizer.from_pretrained(model_name, padding=True, truncation=True, max_length=512)

# Create the pipeline
qa_pipeline = pipeline(
    "question-answering",
    model=model_name,
    tokenizer=tokenizer
)

llm = HuggingFacePipeline(pipeline=qa_pipeline)
```

# 7. Build the Retrieval QA Chain
On connecte le rГ©cupГ©rateur (retriever) au modГЁle de langue.

```python
from langchain.chains import RetrievalQA

retriever = db.as_retriever(search_kwargs={"k": 4})

# Configuration de la chaГ®ne
qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="refine",
    retriever=retriever,
    return_source_documents=False
)
```

# 8. Test your RAG system
Test final avec une question.

```python
question = "What is cheesemaking?"
# Use invoke instead of run in recent LangChain versions
try:
    result = qa.invoke({"query": question})
    print("Question:", question)
    print("Answer:", result["result"])
except Exception as e:
    print(f"An error occurred: {e}")
```
