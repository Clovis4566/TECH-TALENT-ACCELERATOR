### Using a Synthetic Medical Notes Dataset

Since the `sample_notes_data.jsonl` file could not be reliably downloaded from the GitHub repository, we will create a small, synthetic dataset directly in this notebook. This DataFrame `df` will contain sample medical notes to allow you to proceed with the demonstration of embedding and indexing.

```python
# Create a synthetic dataset of medical notes
medical_notes_data = [
    {
        "id": "note_001",
        "text": "Patient presented with mild fever and sore throat. Diagnosed with common cold. Advised rest and hydration."
    },
    {
        "id": "note_002",
        "text": "Follow-up visit for hypertension. Blood pressure 140/90. Medication adjusted. Schedule next appointment in 3 months."
    },
    {
        "id": "note_003",
        "text": "Emergency room visit: Severe abdominal pain. CT scan showed appendicitis. Surgery scheduled for tomorrow morning."
    },
    {
        "id": "note_004",
        "text": "Annual physical exam. All labs normal. Patient in good health. Continue regular exercise."
    },
    {
        "id": "note_005",
        "text": "Patient complains of persistent cough for 2 weeks. Chest X-ray ordered. Prescribed cough suppressants."
    }
]

df = pd.DataFrame(medical_notes_data)

print(f"Number of lines loaded: {len(df)}")
print(df.head())
```

```python
# Show head of the DataFrame
print("Data shape:", df.shape) # Show number of rows and columns
df.head()
```

```python
index = pc.Index(name=index_name)

# Upsert data into index from DataFrame
index.upsert_from_dataframe(df)
```

```python
def is_fresh(index):
    stats = index.describe_index_stats()
    vector_count = stats.total_vector_count
    print(f"Vector count: ", vector_count)
    return vector_count > 0 # What should this be?

while not is_fresh(index):
    time.sleep(5)

print("Index ready!")
index.describe_index_stats()
```

```python
def get_embedding(input_question):
    model_name = 'sentence-transformers/all-MiniLM-L6-v2'
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)

    encoded_input = tokenizer(input_question, padding=True, truncation=True, return_tensors='pt')

    with torch.no_grad():
        model_output = model(**encoded_input)

        # On calcule la moyenne sur la dimension 0 (les tokens)
        # pour obtenir un vecteur unique de taille hidden_size (384)
        embedding = model_output.last_hidden_state[0].mean(dim=0)

    return embedding
```

```python
# Build a query to search
question = "Patient experiencing severe chest pain and shortness of breath"
query = get_embedding(question).tolist()

# Get results
# top_k=3 permet de récupérer les 3 correspondances les plus pertinentes
results = index.query(vector=[query], top_k=3, include_metadata=True)

# Sort results by score in descending order
sorted_matches = sorted(results['matches'], key=lambda x: x['score'], reverse=True)

# Affichage des résultats pour vérification
for i, match in enumerate(sorted_matches):
    print(f"Rang {i+1} [Score: {match['score']:.4f}] : {match['metadata'].get('text', 'Pas de texte disponible')}\n")
```

```python
# Create documents with concatenated metadata field as "reranking_field" field
transformed_documents = [
    {
        'id': match['id'],
        # Remplacé ici par 'metadata'
        'reranking_field': '; '.join([f"{key}: {value}" for key, value in match['metadata'].items()])
    }
    for match in results['matches']
]

# Perform reranking using these transformed_documents
# The reranker expects a 'text' field for the content to be reranked.
# So, we map 'reranking_field' to 'text'.
reranked_medical_notes = pc.inference.rerank(
    model="bge-reranker-v2-m3",
    query=question, # use the medical question
    documents=[{"id": doc['id'], "text": doc['reranking_field']} for doc in transformed_documents],
    top_n=3
)
```

```python
def show_reranked_results(question, matches):
    print(f"Question: '{question}'")
    print('\nReranked Results:')
    for i, match in enumerate(matches):
        print(f'{str(i+1).rjust(4)}. ID: {match.document.id}')
        print(f' Score: {match.score:.4f}') # Attribut du score de reclassement
        print(f' Reranking Field: {match.document.text}') # Now use match.document.text as it was mapped from reranking_field
        print('')

# .data contient la liste des résultats réorganisés
show_reranked_results(question, reranked_medical_notes.data)
```
