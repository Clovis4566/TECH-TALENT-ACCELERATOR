## Single-Head Attention Implementation

Here's the implementation of a single-head attention module in PyTorch, as per your instructions. It includes linear projections for Q, K, and V, calculates attention weights, and handles dummy tensor validation.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SingleHeadAttention(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.hidden_dim = hidden_dim

        # Linear transformations for Q, K, V
        self.query_transform = nn.Linear(hidden_dim, hidden_dim)
        self.key_transform = nn.Linear(hidden_dim, hidden_dim)
        self.value_transform = nn.Linear(hidden_dim, hidden_dim)

        # Scaling factor
        self.scale = torch.sqrt(torch.tensor(hidden_dim, dtype=torch.float32))

    def forward(self, query, key, value, mask=None):
        # query, key, value shape: (batch_size, seq_len, hidden_dim)

        # 1. Apply linear projections
        Q = self.query_transform(query) # (batch_size, seq_len, hidden_dim)
        K = self.key_transform(key)     # (batch_size, seq_len, hidden_dim)
        V = self.value_transform(value) # (batch_size, seq_len, hidden_dim)

        # 2. Calculate attention scores (Q * K^T)
        # (batch_size, seq_len, hidden_dim) @ (batch_size, hidden_dim, seq_len)
        # -> (batch_size, seq_len, seq_len)
        attention_scores = torch.matmul(Q, K.transpose(-2, -1)) / self.scale

        # 3. Apply mask (if provided)
        if mask is not None:
            attention_scores = attention_scores.masked_fill(mask == 0, -1e9) # Fill with a very small number

        # 4. Apply softmax to get attention weights
        attention_weights = F.softmax(attention_scores, dim=-1) # (batch_size, seq_len, seq_len)

        # Log attention weights for inspection
        print("\n--- Attention Weights Sample (First Batch, First Query) ---")
        print(attention_weights[0, 0, :].detach().cpu().numpy()) # Detach and move to CPU for printing
        print("----------------------------------------------------------\n")

        # 5. Multiply weights by V
        # (batch_size, seq_len, seq_len) @ (batch_size, seq_len, hidden_dim)
        # -> (batch_size, seq_len, hidden_dim)
        output = torch.matmul(attention_weights, V)

        return output, attention_weights


# --- Validation with Dummy Tensors ---
print("Validating Single-Head Attention implementation...")
batch_size = 2
seq_len = 5
hidden_dim = 64

# Create dummy tensors
dummy_query = torch.randn(batch_size, seq_len, hidden_dim)
dummy_key = torch.randn(batch_size, seq_len, hidden_dim)
dummy_value = torch.randn(batch_size, seq_len, hidden_dim)

# Instantiate the attention module
attention_module = SingleHeadAttention(hidden_dim)

# Perform forward pass
output, weights = attention_module(dummy_query, dummy_key, dummy_value)

# Validate shapes
print(f"Input Query shape: {dummy_query.shape}")
print(f"Input Key shape: {dummy_key.shape}")
print(f"Input Value shape: {dummy_value.shape}")
print(f"Output shape: {output.shape}")       # Expected: (batch_size, seq_len, hidden_dim)
print(f"Weights shape: {weights.shape}")     # Expected: (batch_size, seq_len, seq_len)

assert output.shape == torch.Size([batch_size, seq_len, hidden_dim])
assert weights.shape == torch.Size([batch_size, seq_len, seq_len])

print("Single-Head Attention validation successful!")
```

## Multi-Head Attention Implementation

To implement Multi-Head Attention, we'll first ensure `einops` is installed, as it provides a clean way to handle tensor reshaping operations. Then, we'll define the `MultiHeadAttention` class, incorporating multiple attention heads, dropout, and residual connections.

```python
# Install einops for cleaner tensor manipulation
!pip install einops
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from einops import rearrange

class MultiHeadAttention(nn.Module):
    def __init__(self, hidden_dim, num_heads, dropout_rate=0.1):
        super().__init__()
        self.hidden_dim = hidden_dim
        self.num_heads = num_heads
        self.head_dim = hidden_dim // num_heads

        assert hidden_dim % num_heads == 0, "hidden_dim must be divisible by num_heads"

        # Linear layers for Q, K, V projections for all heads
        self.query_proj = nn.Linear(hidden_dim, hidden_dim)
        self.key_proj = nn.Linear(hidden_dim, hidden_dim)
        self.value_proj = nn.Linear(hidden_dim, hidden_dim)

        # Final linear layer to combine outputs from all heads
        self.out_proj = nn.Linear(hidden_dim, hidden_dim)

        self.dropout = nn.Dropout(dropout_rate)

        # Scaling factor for attention scores
        self.scale = torch.sqrt(torch.tensor(self.head_dim, dtype=torch.float32))

    def forward(self, query, key, value, mask=None):
        batch_size = query.shape[0]

        # Save original query for residual connection later
        identity = query

        # 1. Apply linear projections for Q, K, V
        Q = self.query_proj(query)
        K = self.key_proj(key)
        V = self.value_proj(value)

        # 2. Reshape to (batch_size, num_heads, seq_len, head_dim)
        # Using einops for cleaner reshaping
        Q = rearrange(Q, 'b s (h d) -> b h s d', h=self.num_heads, d=self.head_dim)
        K = rearrange(K, 'b s (h d) -> b h s d', h=self.num_heads, d=self.head_dim)
        V = rearrange(V, 'b s (h d) -> b h s d', h=self.num_heads, d=self.head_dim)

        # 3. Calculate attention scores for all heads (Q * K^T)
        # (batch_size, num_heads, seq_len, head_dim) @ (batch_size, num_heads, head_dim, seq_len)
        # -> (batch_size, num_heads, seq_len, seq_len)
        attention_scores = torch.matmul(Q, K.transpose(-2, -1)) / self.scale

        # 4. Apply mask (if provided)
        if mask is not None:
            # mask shape is (batch_size, 1, 1, seq_len) or (batch_size, 1, seq_len, seq_len)
            # expand mask to match attention_scores dimensions
            attention_scores = attention_scores.masked_fill(mask == 0, -1e9)

        # 5. Apply softmax to get attention weights
        attention_weights = F.softmax(attention_scores, dim=-1)

        # 6. Apply dropout to weights
        attention_weights = self.dropout(attention_weights)

        # 7. Multiply weights by V
        # (batch_size, num_heads, seq_len, seq_len) @ (batch_size, num_heads, seq_len, head_dim)
        # -> (batch_size, num_heads, seq_len, head_dim)
        attention_output = torch.matmul(attention_weights, V)

        # 8. Concatenate outputs from all heads
        # (batch_size, num_heads, seq_len, head_dim) -> (batch_size, seq_len, num_heads * head_dim)
        attention_output = rearrange(attention_output, 'b h s d -> b s (h d)')

        # 9. Apply final linear projection
        output = self.out_proj(attention_output)

        # 10. Apply residual connection (input + attention_output)
        output = self.dropout(output) + identity

        return output, attention_weights


# --- Validation with Dummy Tensors ---
print("\nValidating Multi-Head Attention implementation...")
batch_size = 2
seq_len = 10
hidden_dim = 256 # Must be divisible by num_heads
num_heads = 8
dropout_rate = 0.1

# Create dummy tensors
dummy_query = torch.randn(batch_size, seq_len, hidden_dim)
dummy_key = torch.randn(batch_size, seq_len, hidden_dim)
dummy_value = torch.randn(batch_size, seq_len, hidden_dim)

# Create a dummy mask (e.g., for padding or causal attention)
# mask = torch.ones(batch_size, 1, seq_len, seq_len) # Example: no masking
# For simplicity, let's use a mask that allows all tokens to attend to each other initially
# but you might want to create a more realistic mask for actual use cases
# Example: mask = torch.randint(0, 2, (batch_size, 1, 1, seq_len)) # Simple sequence mask

# Instantiate the MultiHeadAttention module
multi_head_attention = MultiHeadAttention(hidden_dim, num_heads, dropout_rate)

# Perform forward pass
output, weights = multi_head_attention(dummy_query, dummy_key, dummy_value)

# Validate shapes
print(f"Input Query shape: {dummy_query.shape}")
print(f"Input Key shape: {dummy_key.shape}")
print(f"Input Value shape: {dummy_value.shape}")
print(f"Output shape: {output.shape}")       # Expected: (batch_size, seq_len, hidden_dim)
print(f"Weights shape: {weights.shape}")     # Expected: (batch_size, num_heads, seq_len, seq_len)

assert output.shape == torch.Size([batch_size, seq_len, hidden_dim])
assert weights.shape == torch.Size([batch_size, num_heads, seq_len, seq_len])

print("Multi-Head Attention validation successful!")
```

## Custom Encoder Stack Implementation

This section will define the building blocks for an encoder-only Transformer network: a position-wise Feed-Forward Network (FFN), an Encoder Block combining Multi-Head Attention with FFN and Layer Normalization, and finally, a full Transformer Encoder by stacking these blocks.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# Helper function for residual connection and layer normalization
class LayerNorm(nn.Module):
    def __init__(self, features, eps=1e-6):
        super(LayerNorm, self).__init__()
        self.a_2 = nn.Parameter(torch.ones(features))
        self.b_2 = nn.Parameter(torch.zeros(features))
        self.eps = eps

    def forward(self, x):
        mean = x.mean(-1, keepdim=True)
        std = x.std(-1, keepdim=True)
        return self.a_2 * (x - mean) / (std + self.eps) + self.b_2

# Position-wise Feed-Forward Network
class FeedForward(nn.Module):
    def __init__(self, hidden_dim, ff_dim, dropout_rate=0.1):
        super().__init__()
        self.linear_1 = nn.Linear(hidden_dim, ff_dim)
        self.dropout = nn.Dropout(dropout_rate)
        self.linear_2 = nn.Linear(ff_dim, hidden_dim)

    def forward(self, x):
        # (batch_size, seq_len, hidden_dim) -> (batch_size, seq_len, ff_dim) -> (batch_size, seq_len, hidden_dim)
        return self.linear_2(self.dropout(F.relu(self.linear_1(x))))

# Transformer Encoder Block
class EncoderBlock(nn.Module):
    def __init__(self, hidden_dim, num_heads, ff_dim, dropout_rate=0.1):
        super().__init__()
        self.attention = MultiHeadAttention(hidden_dim, num_heads, dropout_rate)
        self.feed_forward = FeedForward(hidden_dim, ff_dim, dropout_rate)

        self.norm1 = LayerNorm(hidden_dim)
        self.norm2 = LayerNorm(hidden_dim)
        self.dropout1 = nn.Dropout(dropout_rate)
        self.dropout2 = nn.Dropout(dropout_rate)

    def forward(self, x, mask=None):
        # x shape: (batch_size, seq_len, hidden_dim)

        # Multi-Head Attention with residual connection and layer norm
        # The MultiHeadAttention already includes a residual connection and dropout in its return 'output', so we adjust here.
        # The identity for the attention is handled inside MultiHeadAttention. Here, we just apply norm.
        normed_x = self.norm1(x)
        attn_output, _ = self.attention(normed_x, normed_x, normed_x, mask) # Self-attention
        x = x + self.dropout1(attn_output) # Residual connection for attention

        # Feed-Forward Network with residual connection and layer norm
        normed_x = self.norm2(x)
        ff_output = self.feed_forward(normed_x)
        x = x + self.dropout2(ff_output) # Residual connection for FFN

        return x

# Full Transformer Encoder
class TransformerEncoder(nn.Module):
    def __init__(self, num_layers, hidden_dim, num_heads, ff_dim, dropout_rate=0.1):
        super().__init__()
        self.layers = nn.ModuleList([
            EncoderBlock(hidden_dim, num_heads, ff_dim, dropout_rate)
            for _ in range(num_layers)
        ])
        self.norm = LayerNorm(hidden_dim)

    def forward(self, x, mask=None):
        for layer in self.layers:
            x = layer(x, mask)
        return self.norm(x)


# --- Validation with Dummy Tensors for Encoder Stack ---
print("\nValidating Transformer Encoder stack...")

batch_size = 2
seq_len = 10
hidden_dim = 256
num_heads = 8
ff_dim = 512 # Feed forward hidden dimension
num_layers = 3
dropout_rate = 0.1

# Create dummy input tensor
dummy_input = torch.randn(batch_size, seq_len, hidden_dim)

# Instantiate and pass through the Transformer Encoder
transformer_encoder = TransformerEncoder(num_layers, hidden_dim, num_heads, ff_dim, dropout_rate)
encoder_output = transformer_encoder(dummy_input)

# Validate output shape
print(f"Input shape: {dummy_input.shape}")
print(f"Encoder output shape: {encoder_output.shape}") # Expected: (batch_size, seq_len, hidden_dim)

assert encoder_output.shape == torch.Size([batch_size, seq_len, hidden_dim])
print("Transformer Encoder validation successful!")
```

## Dataset Preparation: Tokenization and DataLoaders

We will use the `transformers` library to tokenize our text data (premise and hypothesis). We'll concatenate them, encode them into input IDs, attention masks, and token type IDs, and then create PyTorch datasets and dataloaders for efficient training.

```python
# Install the transformers library
!pip install transformers datasets
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer
import pandas as pd

# Assuming df_train and df_test are already loaded from previous steps
# For demonstration, let's ensure they are defined if this cell is run independently
if 'df_train' not in locals():
    print("df_train not found. Loading dummy data for demonstration.")
    data = {
        'premise': ["The cat sat on the mat.", "A person is walking their dog.", "I enjoy reading books."],
        'hypothesis': ["The cat is comfortable.", "The dog is on a leash.", "Books are a great escape."],
        'lang_abv': ["en", "en", "en"],
        'label': [0, 1, 2]
    }
    df_train = pd.DataFrame(data)
    df_test = pd.DataFrame(data)

# Define the tokenizer
# We'll use a multilingual BERT tokenizer, suitable for the NLI dataset
# You might choose a different tokenizer based on your specific model architecture
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")

# Define maximum sequence length
MAX_LEN = 128 # A common choice for BERT-based models

class NLIDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, index):
        row = self.dataframe.iloc[index]
        premise = str(row['premise'])
        hypothesis = str(row['hypothesis'])
        label = row['label']

        # Tokenize premise and hypothesis
        # Using truncation=True and padding='max_length' to handle variable lengths
        encoding = self.tokenizer.encode_plus(
            premise,
            hypothesis,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=True,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'token_type_ids': encoding['token_type_ids'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# Create dataset instances
train_dataset = NLIDataset(df_train, tokenizer, MAX_LEN)
test_dataset = NLIDataset(df_test, tokenizer, MAX_LEN)

# Define DataLoader parameters
BATCH_SIZE = 16 # Adjust based on your GPU memory

# Create DataLoader instances
train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"\nTrain dataset size: {len(train_dataset)}")
print(f"Test dataset size: {len(test_dataset)}")
print(f"Number of train batches: {len(train_dataloader)}")
print(f"Number of test batches: {len(test_dataloader)}")

# Example of one batch
for batch in train_dataloader:
    print("\nExample batch shapes:")
    print(f"Input IDs shape: {batch['input_ids'].shape}")
    print(f"Attention Mask shape: {batch['attention_mask'].shape}")
    print(f"Token Type IDs shape: {batch['token_type_ids'].shape}")
    print(f"Labels shape: {batch['labels'].shape}")
    break

print("Dataset and DataLoaders are ready!")
```
