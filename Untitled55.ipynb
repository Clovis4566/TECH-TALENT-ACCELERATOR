```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# ==========================================
# 1. TÉLÉCHARGEMENT ET CHARGEMENT DES DONNÉES
# ==========================================
print("Veuillez sélectionner votre fichier CSV :")
uploaded = files.upload()

# Récupération du nom du fichier chargé
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print("\n--- Aperçu des premières lignes ---")
print(df.head())

# ==========================================
# 2. TRAITEMENT DES VALEURS MANQUANTES
# ==========================================
print("\n--- Vérification des valeurs manquantes ---")
print(df.isnull().sum())

# Remplacement des valeurs manquantes (imputation par la médiane pour le numérique)
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Pour les colonnes catégorielles, on remplace par le mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# ==========================================
# 3. SUPPRESSION DES COLONNES INUTILES
# ==========================================
# Supposons qu'il existe une colonne 'id' ou 'Unnamed: 32' fréquente dans certains datasets
columns_to_drop = ['id', 'Unnamed: 32']
for col in columns_to_drop:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)
        print(f"\nColonne inutile '{col}' supprimée.")

# ==========================================
# 4. GRAPHIQUE DE COMPTAGE (DIAGNOSTIC)
# ==========================================
# Note : Remplacer 'diagnostic' par le nom exact de votre colonne si la casse diffère (ex: 'Diagnosis')
target_col = 'diagnostic' if 'diagnostic' in df.columns else 'diagnosis'

if target_col in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=target_col, palette='viridis')
    plt.title(f"Distribution de la colonne : {target_col}")
    plt.xlabel(target_col)
    plt.ylabel("Nombre de cas")
    plt.show()
else:
    print(f"\nAttention : La colonne de diagnostic n'a pas été trouvée avec le nom attendu.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Identifions à nouveau la variable cible
target_col = 'diagnostic' if 'diagnostic' in df.columns else 'diagnosis'

# ==========================================
# 1. LIGNES UNIQUES & ENCODAGE CATÉGORIEL
# ==========================================
print(f"Nombre de lignes uniques dans '{target_col}' : {df[target_col].nunique()}")
print("Valeurs uniques :", df[target_col].unique())

# Encodage de la variable cible en valeurs numériques (ex: M/B -> 1/0)
le = LabelEncoder()
df[target_col] = le.fit_transform(df[target_col])

# Encodage des autres variables catégorielles si elles existent
for col in df.select_dtypes(include=['object']).columns:
    df[col] = LabelEncoder().fit_transform(df[col])

# Separation des fonctionnalités (X) et de la cible (y)
X = df.drop(columns=[target_col])
y = df[target_col]

# NORMAlISATION : Crucial pour KNN et SVM
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==========================================
# 2. DIVISION DES DONNÉES (TRAIN / TEST)
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(f"\nDonnées divisées : {X_train.shape[0]} entraînement, {X_test.shape[0]} test.")

# Dictionnaire pour stocker les scores de précision
scores = {}

# ==========================================
# 3. ENTRAÎNEMENT ET ÉVALUATION DES MODÈLES
# ==========================================

# A. Régression Logistique
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
scores['Régression Logistique'] = accuracy_score(y_test, lr.predict(X_test))

# B. K plus proches voisins (KNN)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
scores['KNN'] = accuracy_score(y_test, knn.predict(X_test))

# C. Forêts Aléatoires (Random Forest)
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
scores['Forêts Aléatoires'] = accuracy_score(y_test, rf.predict(X_test))

# D. Machine à Vecteurs de Support (SVM)
svm = SVC()
svm.fit(X_train, y_train)
scores['SVM'] = accuracy_score(y_test, svm.predict(X_test))

# ==========================================
# 4. COMPARAISON ET MEILLEUR MODÈLE
# ==========================================
print("\n================ COMPARAISON DES MODÈLES ================")
df_scores = pd.DataFrame(list(scores.items()), columns=['Modèle', 'Précision (Accuracy)'])
print(df_scores.to_string(index=False))
print("=========================================================")

best_model = max(scores, key=scores.get)
print(f"\nLe meilleur modèle est : ** {best_model} ** avec une précision de {scores[best_model]:.4f}")
```
