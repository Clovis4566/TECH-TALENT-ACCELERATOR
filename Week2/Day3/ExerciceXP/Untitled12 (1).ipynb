```python
# Cellule 1 : Importation des bibliothèques nécessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats

print("Toutes les bibliothèques sont chargées avec succès !")
#Exercice 1
print("La version de scipy est :",scipy.__version__)
#Exercice 2
# la donner de notre exercice est
data = [12, 15, 13, 12, 18, 20, 22, 21]
moyenne = np.mean(data)
mediane = np.median(data)
variance = np.var(data,ddof=1)
ecart_type=np.std(data,ddof=1)
# Afficher les résultats
print(f"Moyenne : {moyenne}")
print(f"Médiane : {mediane}")
print(f"Variance : {variance:.2f}")
print(f"Écart-type : {ecart_type:.2f}")
# Exercice 3 corrigé : Génération et visualisation d'une distribution normale
from scipy.stats import norm

# Paramètres de la distribution
moyenne_dist = 50
ecart_type_dist = 10

# Génération de points pour l'axe X
x = np.linspace(moyenne_dist - 4 * ecart_type_dist, moyenne_dist + 4 * ecart_type_dist, 1000)
# Calcul de la Densité de Probabilité (PDF)
y = norm.pdf(x, moyenne_dist, ecart_type_dist)

# Tracé du graphique
plt.figure(figsize=(8, 5))
# Le 'rf' ici permet d'utiliser à la fois le formatage de variables {} et le texte brut pour LaTeX (\mu, \sigma)
plt.plot(x, y, label=rf"Normale ($\mu={moyenne_dist}$, $\sigma={ecart_type_dist}$)", color="blue", lw=2)
plt.title("Distribution Normale Spécifiée")
plt.xlabel("Valeurs")
plt.ylabel("Densité de probabilité")
plt.axvline(moyenne_dist, color="red", linestyle="--", label="Moyenne (50)")
plt.legend()
plt.grid(True)
plt.show()

# Bonus : Tout avoir d'un coup avec SciPy
description = stats.describe(data)
print("\nDescription globale avec SciPy :", description)

# Exercice 4 : Test T indépendant entre deux échantillons
np.random.seed(42) # Pour avoir les mêmes résultats à chaque exécution

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Exécution du test t de Student (échantillons indépendants)
t_stat, p_value = stats.ttest_ind(data1, data2)

print(f"Statistique T : {t_stat:.4f}")
print(f"Valeur p (p-value) : {p_value:.4e}")

# Interprétation automatique
alpha = 0.05
if p_value < alpha:
    print("\nRésultat : On rejette l'hypothèse nulle ($H_0$). Les moyennes des deux groupes sont significativement différentes.")
else:
    print("\nRésultat : On ne peut pas rejeter l'hypothèse nulle ($H_0$). Il n'y a pas de différence significative entre les moyennes.")
    # Exercice 5 : Régression linéaire pour l'immobilier
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]

# Calcul de la régression linéaire
pente, ordonnee_origine, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print(f"1. Pente (Slope) : {pente:.2f}")
print(f"   Ordonnée à l'origine (Intercept) : {ordonnee_origine:.2f}")

# 2. Estimation pour une maison de 90 m²
taille_cible = 90
prix_predit = (pente * taille_cible) + ordonnee_origine
print(f"\n2. Le prix estimé pour une maison de 90 m² est de : {prix_predit:,.2f} unités monétaires.")

# 3. Interprétation textuelle
print(f"\n3. Interprétation de la pente : Chaque mètre carré supplémentaire ajoute en moyenne {pente:.2f} unités monétaires au prix de la maison.")
# Exercice  : Analyse de la variance (ANOVA) pour les engrais
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Exécution de l'ANOVA à un facteur (One-way ANOVA)
f_stat, p_value_anova = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"1. Valeur F : {f_stat:.4f}")
print(f"   Valeur p : {p_value_anova:.6f}")

# 2. Réponses aux questions de compréhension
print("\n2. Analyse de la valeur p :")
if p_value_anova < 0.05:
    print("   👉 Les engrais ont des effets significativement différents sur la croissance des plantes (p < 0.05).")
else:
    print("   👉 Pas de différence statistiquement significative observée.")

print("\n3. Que se passerait-il si la valeur p était supérieure à 0.05 ?")
print("   👉 Si p > 0.05, cela signifierait que nous n'aurions pas de preuves statistiques suffisantes pour affirmer que les engrais ont des impacts différents. Les écarts observés entre les groupes pourraient simplement être dus au hasard.")
# Exercice 7 : Distribution Binomiale (Lancers de pièce)
from scipy.stats import binom

n = 10  # Nombre de lancers
p = 0.5 # Probabilité d'obtenir "Face" sur un lancer équilibré
k = 5   # Nombre exact de succès recherchés

# Calcul de la Probabilité de Masse (PMF)
probabilite_5_faces = binom.pmf(k, n, p)

print(f"La probabilité d'obtenir exactement 5 faces en 10 lancers est de : {probabilite_5_faces * 100:.2f}%")
# Exercice 8 : Calcul des corrélations de Pearson et Spearman
data_df = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Corrélation de Pearson (relation linéaire)
pearson_corr, p_val_p = stats.pearsonr(data_df['age'], data_df['income'])

# Corrélation de Spearman (relation monotone / rangs)
spearman_corr, p_val_s = stats.spearmanr(data_df['age'], data_df['income'])

print(f"Coefficient de corrélation de Pearson : {pearson_corr:.4f} (p-value: {p_val_p})")
print(f"Coefficient de corrélation de Spearman : {spearman_corr:.4f} (p-value: {p_val_s})")
print("\nAnalyse : On observe une corrélation positive parfaite (1.0) entre l'âge et le revenu sur cet échantillon.")
```
