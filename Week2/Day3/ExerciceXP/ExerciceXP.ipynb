```python

# =====================================================================
# TECH TALENT ACCELERATOR - SEMAIN 2 - JOUR 3 : EXERCICES XP
# =====================================================================

# ---------------------------------------------------------------------
# EXERCICE 1 : Utilisation de base de SciPy
# ---------------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy import stats
from scipy.stats import norm, binom

print(f"EXERCICE 1 : La version de SciPy est {scipy.__version__}")


# ---------------------------------------------------------------------
# EXERCICE 2 : Statistiques descriptives
# ---------------------------------------------------------------------
data_xp = [12, 15, 13, 12, 18, 20, 22, 21]

moyenne = np.mean(data_xp)
mediane = np.median(data_xp)
variance = np.var(data_xp, ddof=1)
ecart_type = np.std(data_xp, ddof=1)

print("\nEXERCICE 2 : Statistiques descriptives")
print(f"Moyenne : {moyenne}")
print(f"Mediane : {mediane}")
print(f"Variance : {variance:.2f}")
print(f"Ecart-type : {ecart_type:.2f}")


# ---------------------------------------------------------------------
# EXERCICE 3 : Comprendre les distributions
# ---------------------------------------------------------------------
moyenne_dist = 50
ecart_type_dist = 10

# Generation des points de l'axe X et de la courbe
x = np.linspace(moyenne_dist - 4 * ecart_type_dist, moyenne_dist + 4 * ecart_type_dist, 1000)
y = norm.pdf(x, moyenne_dist, ecart_type_dist)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color="blue", lw=2, label="Distribution Normale (50, 10)")
plt.title("Exercice 3 : Distribution Normale")
plt.axvline(moyenne_dist, color="red", linestyle="--", label="Moyenne (50)")
plt.legend()
plt.show()
print("EXERCICE 3 : Graphique de la distribution normale genere.")


# ---------------------------------------------------------------------
# EXERCICE 4 : Application du test t
# ---------------------------------------------------------------------
np.random.seed(42)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value_t = stats.ttest_ind(data1, data2)

print("\nEXERCICE 4 : Test T de Student")
print(f"Statistique T : {t_stat:.4f}")
print(f"Valeur p : {p_value_t:.4e}")
if p_value_t < 0.05:
    print("Interpretation : On rejette H0, les moyennes sont significativement differentes.")
else:
    print("Interpretation : On ne peut pas rejeter H0, pas de difference significative.")


# ---------------------------------------------------------------------
# EXERCICE 5 : Analyse de regression lineaire
# ---------------------------------------------------------------------
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]

pente, ordonnee, r_val, p_val_reg, std_err = stats.linregress(house_sizes, house_prices)
prix_90m2 = (pente * 90) + ordonnee

print("\nEXERCICE 5 : Regression lineaire pour l'immobilier")
print(f"1. Pente : {pente:.2f} | Ordonnee a l'origine : {ordonnee:.2f}")
print(f"2. Prix estime pour une maison de 90 m2 : {prix_90m2:,.2f} unites monetaires.")
print(f"3. Signification : Chaque m2 supplementaire ajoute en moyenne {pente:.2f} unites au prix.")


# ---------------------------------------------------------------------
# EXERCICE 6 : Comprendre l'ANOVA
# ---------------------------------------------------------------------
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_stat, p_value_anova = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("\nEXERCICE 6 : Analyse de la Variance (ANOVA)")
print(f"1. Valeur F : {f_stat:.4f} | Valeur p : {p_value_anova:.6f}")
if p_value_anova < 0.05:
    print("2. Conclusion : Les engrais ont des effets significativement differents (p < 0.05).")
else:
    print("2. Conclusion : Aucune difference significative detectee.")
print("3. Si p > 0.05 : Les ecarts observes seraient dus uniquement au hasard.")


# ---------------------------------------------------------------------
# EXERCICE 7 : Distributions de probabilite (Optionnel)
# ---------------------------------------------------------------------
prob_5_faces = binom.pmf(5, 10, 0.5)
print("\nEXERCICE 7 : Distribution Binomiale")
print(f"Probabilite d'avoir exactement 5 faces sur 10 lancers : {prob_5_faces * 100:.2f}%")


# ---------------------------------------------------------------------
# EXERCICE 8 : Coefficients de correlation (Optionnel)
# ---------------------------------------------------------------------
df_corr = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

pearson_c, _ = stats.pearsonr(df_corr['age'], df_corr['income'])
spearman_c, _ = stats.spearmanr(df_corr['age'], df_corr['income'])

print("\nEXERCICE 8 : Coefficients de correlation")
print(f"Correlation de Pearson : {pearson_c:.4f}")
print(f"Correlation de Spearman : {spearman_c:.4f}")
print("Analyse : Il existe une relation positive parfaite (1.0) entre l'age et le revenu.")

print("\n=====================================================================")
print("--- TOUS LES EXERCICES XP ONT ETE EXECUTES AVEC SUCCES ---")
print("=====================================================================")
```
