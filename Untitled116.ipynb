## 1. Data Loading and Generators

This section contains the pre-filled code for loading the image data, building dataframes from the image folders, and setting up Keras `ImageDataGenerator`s for training, validation, and testing. It also performs a train-validation split and prints a summary of the dataset. Please ensure you have downloaded and extracted the `cats_dogs` dataset into a `data/cats_dogs` folder relative to your notebook's location.

```python
import os, math, re, random
from glob import glob
from pathlib import Path
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split

np.random.seed(42); tf.random.set_seed(42)

# Paths - change if needed
DATA_ROOT = Path("data/cats_dogs")
train_dir = (DATA_ROOT / "train" / "train") if (DATA_ROOT / "train" / "train").exists() else (DATA_ROOT / "train")
test_dir  = (DATA_ROOT / "test"  / "test")  if (DATA_ROOT / "test"  / "test").exists()  else (DATA_ROOT / "test")

IMG_HEIGHT, IMG_WIDTH = 180, 180
batch_size = 32
seed = 1337

# Build DataFrames from folders
def build_df_from_folder(folder: Path, labeled: bool=True):
    exts = ('*.jpg','*.jpeg','*.png','*.bmp')
    files = []
    for ex in exts:
        files.extend(glob(str(folder / '**' / ex), recursive=True))
    if not files:
        raise FileNotFoundError(f"No images found under {folder}")
    rows = []
    for f in files:
        if labeled:
            name = Path(f).name.lower()
            parent = Path(f).parent.name.lower()
            if parent in {"cat","cats"}:
                label = "cat"
            elif parent in {"dog","dogs"}:
                label = "dog"
            else:
                if re.search(r'(^|[^a-z])cat([^a-z]|$)', name): label = "cat"
                elif re.search(r'(^|[^a-z])dog([^a-z]|$)', name): label = "dog"
                else:
                    continue
            rows.append({"filepath": f, "label": label})
        else:
            rows.append({"filepath": f})
    return pd.DataFrame(rows)

df_train_full = build_df_from_folder(train_dir, labeled=True)
df_test_full  = build_df_from_folder(test_dir,  labeled=False)

# Train validation split
df_tr, df_val = train_test_split(
    df_train_full, test_size=0.2, stratify=df_train_full["label"], random_state=seed
)

# Generators
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.5,
    horizontal_flip=True,
)
val_gen = ImageDataGenerator(rescale=1./255)
test_gen = ImageDataGenerator(rescale=1./255)

train_flow = train_gen.flow_from_dataframe(
    df_tr, x_col="filepath", y_col="label",
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    class_mode="binary", batch_size=batch_size,
    shuffle=True, seed=seed, validate_filenames=False
)
val_flow = val_gen.flow_from_dataframe(
    df_val, x_col="filepath", y_col="label",
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    class_mode="binary", batch_size=batch_size,
    shuffle=False, validate_filenames=False
)
# Unlabeled test for inference only
test_flow = test_gen.flow_from_dataframe(
    df_test_full, x_col="filepath", y_col=None,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    class_mode=None, batch_size=batch_size,
    shuffle=False, validate_filenames=False
)

print({"train": train_flow.samples, "val": val_flow.samples, "test": test_flow.samples,
       "class_indices": train_flow.class_indices})
```

## 2. Inspect the Data

Now that the data generators are set up, let's inspect the data. This involves understanding the class distribution, checking for balance, and visualizing some image samples to identify potential sources of variability.


```python
# Get class indices and labels from train_flow
class_indices = train_flow.class_indices
labels = train_flow.labels

# Calculate total number of training images per class
class_counts = pd.Series(labels).map({v: k for k, v in class_indices.items()}).value_counts()

print("Total number of training images per class:")
print(class_counts)

print(f"\nClass indices: {class_indices}")

# Determine if classes are balanced
if class_counts.max() / class_counts.min() < 1.2: # A simple heuristic for 'balanced'
    balance_status = "balanced"
else:
    balance_status = "imbalanced"

print(f"\nThe classes are {balance_status}.")

print("\nPotential sources of visual variability in cat and dog images can include:")
print("- **Pose**: Animals can be captured in various positions (sitting, standing, running, sleeping).")
print("- **Scale**: Images might show the animal up close or far away, leading to varying sizes in the frame.")
print("- **Lighting**: Different lighting conditions (bright sun, shade, indoor lighting, night with flash) can significantly alter appearance.")
print("- **Background**: Animals can be in diverse environments (grass, concrete, furniture, indoors, outdoors), which can add clutter.")
print("- **Breed Variation**: Both cats and dogs have many breeds with very different appearances (e.g., long-haired vs. short-haired, different ear shapes, sizes, colors).")
print("- **Occlusion**: Parts of the animal might be hidden by objects or other animals.")
```

### Plotting Sample Images

To further inspect the data, let's visualize a small grid of training images and annotate each with its corresponding label. This helps in understanding the visual cues that a model might use to distinguish between cats and dogs.

```python
import matplotlib.pyplot as plt

# Get a batch of images and labels from the training flow
images, labels_batch = next(train_flow)

# Convert binary labels back to 'cat' or 'dog' for display
label_map = {v: k for k, v in train_flow.class_indices.items()}
predicted_labels = np.vectorize(label_map.get)(labels_batch)

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10, 10))
axes = axes.flatten()

# Plot each image
for i, (image, label) in enumerate(zip(images, predicted_labels)):
    if i >= len(axes): # Ensure we don't go out of bounds for axes
        break
    axes[i].imshow(image)
    axes[i].set_title(f"Label: {label}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()

print("\nVisual cues that might help the model distinguish cats from dogs include:")
print("- **Facial features**: Cats typically have more pointed ears, smaller noses, and different eye shapes compared to dogs.")
print("- **Body shape and size**: While variable, dogs often appear more robust or muscular, and their overall body structure can differ.")
print("- **Fur patterns and texture**: Some distinct fur patterns are more common in one species than the other.")
print("- **Tail position and shape**: The way tails are held or their typical shape can sometimes be a differentiator.")
print("- **Behavioral cues**: Though less common in static images, certain poses or actions can be indicative (e.g., dog fetching, cat stretching).")
```
