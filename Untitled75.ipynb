# Challenge : Multi-Head Attention & Transformers

### 💡 Concept d'entretien : Pourquoi l'Attention ?
En entretien, expliquez que l'attention remplace les RNNs en permettant un accès direct à n'importe quel mot de la séquence, peu importe sa distance, résolvant ainsi le problème de disparition du gradient et permettant la parallélisation.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math

class SingleHeadAttention(nn.Module):
    def __init__(self, embed_dim):
        super().__init__()
        self.embed_dim = embed_dim

        # Projections linéaires pour Query, Key, et Value
        self.q_linear = nn.Linear(embed_dim, embed_dim)
        self.k_linear = nn.Linear(embed_dim, embed_dim)
        self.v_linear = nn.Linear(embed_dim, embed_dim)

    def forward(self, x):
        batch_size, seq_len, embed_dim = x.size()

        # 1. Générer Q, K, V
        q = self.q_linear(x)
        k = self.k_linear(x)
        v = self.v_linear(x)

        # 2. Scaled Dot-Product Attention
        # Score = (Q * K^T) / sqrt(d_k)
        scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(embed_dim)

        # 3. Softmax pour obtenir les poids d'attention (somme = 1)
        attn_weights = F.softmax(scores, dim=-1)

        # 4. Multiplication par V
        output = torch.matmul(attn_weights, v)

        return output, attn_weights

# Validation
embed_dim = 64
model_sh = SingleHeadAttention(embed_dim)
dummy_input = torch.randn(1, 10, embed_dim) # (batch, seq, dim)
out, weights = model_sh(dummy_input)
print(f"Shape sortie Single-Head: {out.shape}")
```

### 🚀 Multi-Head Attention (MHA)
**Question d'entretien : Pourquoi plusieurs têtes ?**
Réponse : Une seule tête calcule une moyenne. Plusieurs têtes permettent au modèle de regarder simultanément différentes relations (ex: une tête pour la grammaire, une autre pour le contexte temporel).

```python
class MultiHeadAttention(nn.Module):
    def __init__(self, embed_dim, num_heads, dropout=0.1):
        super().__init__()
        assert embed_dim % num_heads == 0, "embed_dim doit être divisible par num_heads"

        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads

        self.q_linear = nn.Linear(embed_dim, embed_dim)
        self.k_linear = nn.Linear(embed_dim, embed_dim)
        self.v_linear = nn.Linear(embed_dim, embed_dim)

        self.out_proj = nn.Linear(embed_dim, embed_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        batch_size, seq_len, embed_dim = x.size()

        # Projection et split en têtes
        q = self.q_linear(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        k = self.k_linear(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        v = self.v_linear(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)

        # Scaled dot-product
        scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(self.head_dim)
        attn = F.softmax(scores, dim=-1)

        # Concaténation des têtes
        context = torch.matmul(attn, v).transpose(1, 2).contiguous().view(batch_size, seq_len, embed_dim)

        return self.out_proj(context)

# Test MHA
mha = MultiHeadAttention(embed_dim=128, num_heads=8)
dummy_x = torch.randn(8, 20, 128)
out_mha = mha(dummy_x)
print(f"Shape sortie Multi-Head: {out_mha.shape}")
```

### 🧱 Le Bloc Encoder Complet
Un Transformer n'est pas juste de l'attention. C'est un empilement de blocs contenant :
1. Multi-Head Attention
2. Layer Norm & Residual Connection
3. Feed-Forward Network (deux couches linéaires avec ReLU)
4. Layer Norm & Residual Connection

```python
class TransformerEncoderBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout=0.1):
        super().__init__()
        self.mha = MultiHeadAttention(embed_dim, num_heads, dropout)
        self.norm1 = nn.LayerNorm(embed_dim)
        self.norm2 = nn.LayerNorm(embed_dim)

        self.ff = nn.Sequential(
            nn.Linear(embed_dim, ff_dim),
            nn.ReLU(),
            nn.Linear(ff_dim, embed_dim)
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        # 1. Attention + Residual + Norm
        attn_out = self.mha(x)
        x = self.norm1(x + self.dropout(attn_out))

        # 2. Feed Forward + Residual + Norm
        ff_out = self.ff(x)
        x = self.norm2(x + self.dropout(ff_out))
        return x

# Test du bloc complet
encoder_block = TransformerEncoderBlock(embed_dim=128, num_heads=8, ff_dim=512)
output = encoder_block(dummy_x)
print(f"Shape sortie bloc encoder: {output.shape}")
```

### ⚖️ Comparaison : Custom vs Pretrained (BERT)
**Point d'entretien : Pourquoi utiliser BERT plutôt que votre code ?**

| Caractéristique | Votre Module Custom | Pretrained (BERT/DistilBERT) |
| :--- | :--- | :--- |
| **Poids** | Aléatoires (doit apprendre de zéro) | Pré-entraînés sur des Go de texte |
| **Usage** | Tâches spécifiques / niche | Compréhension générale du langage |
| **Coût** | Faible (léger) | Élevé (mémoire/calcul) |

### 📊 Visualisation des poids d'Attention
En entretien, si on vous demande : "Comment interpréter un Transformer ?", parlez des **Attention Maps**.

```python
import matplotlib.pyplot as plt

def plot_attention(weights, tokens):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1)
    cax = ax.matshow(weights[0].detach().numpy(), cmap='viridis')

    ax.set_xticks(range(len(tokens)))
    ax.set_yticks(range(len(tokens)))
    ax.set_xticklabels(tokens, rotation=90)
    ax.set_yticklabels(tokens)

    plt.title("Attention Weights Matrix")
    plt.show()

# Exemple visuel avec notre SingleHeadAttention
sample_text = "The cat sat on the mat".split()
model_sh_viz = SingleHeadAttention(embed_dim=128)
dummy_seq = torch.randn(1, len(sample_text), 128)
_, weights = model_sh_viz(dummy_seq)

plot_attention(weights, sample_text)
```

### 🏁 Résumé pour l'entretien

1.  **Pourquoi le scaling (sqrt(d_k))?** Pour garder la variance à 1. Sans cela, le softmax a des gradients quasi nuls.
2.  **Multi-head vs Single-head** : Le multi-head permet de capturer plusieurs types de dépendances (ex: syntaxique vs sémantique) en parallèle.
3.  **Embeddings vs Attention** : Les embeddings donnent le sens d'un mot seul, l'attention donne son sens *dans son contexte actuel*.
